```python
"""
Contexto del dataset
────────────────────
Variable dependiente: precio_vivienda (€)
Variables explicativas:
  x1 = superficie (m²)          → continua, independiente
  x2 = habitaciones             → DERIVADA de x1 (multicolineal)
  x3 = distancia_centro (km)    → continua, independiente

Cada sección viola UN supuesto diferente al diseñar y residual.

"""
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Librería SciPy (Scientific Python) es una librería de computación científica.
# El módulo stats específicamente contiene herramientas estadísticas:

# Tests de hipótesis (t-test, chi-cuadrado, Shapiro-Wilk...)
# Distribuciones de probabilidad (normal, t de Student, F...)
# Correlaciones y regresiones
# Tests de normalidad
# Entre otros
from scipy import stats

# warnings es un módulo estándar de Python para gestionar mensajes de advertencia.
# filterwarnings("ignore") simplemente silencia todos los warnings durante la ejecución
# Úsalo con cuidado — los warnings a veces avisan de problemas reales en el modelo.
# En producción es mejor gestionarlos que ignorarlos; en notebooks exploratorios
# es habitual silenciarlos para mantener la salida limpia.
import warnings
warnings.filterwarnings("ignore")

# Statsmodels es una librería especializada en estadística inferencial y econometría:
# Modelos de regresión:
    # sm.OLS(y, X).fit()        # Regresión lineal (Ordinary Least Squares)
    # sm.Logit(y, X).fit()      # Regresión logística
    # sm.GLS(y, X).fit()        # Mínimos cuadrados generalizados
# Series temporales:
    #sm.tsa.ARIMA(y, order=(1,1,1)).fit()   # Modelo ARIMA
    #sm.tsa.seasonal_decompose(y)           # Descomposición estacional
# Tests estadísticos:
    # sm.stats.durbin_watson(residuos)       # Test de autocorrelación
    # sm.stats.het_breuschpagan(res, X)      # Test de homocedasticidad
# Gráficos diagnósticos:
    # sm.qqplot(residuos, line="s")          # Q-Q plot ← lo que usamos nosotros
    # sm.graphics.plot_acf(residuos)         # Autocorrelación
import statsmodels.api as sm

# La función variance_inflation_factor
# (Es una métrica que mide cuánto se infla la varianza de un coeficiente β debido a su correlación con las demás variables.)
# Pertenece al submodulo de outliers_influence del módulo de estadística de statmodels
from statsmodels.stats.outliers_influence import variance_inflation_factor

# ¿Qué es sklearn (scikit-learn)?
# Es una de las librerías de Machine Learning más usada en Python.
# Contiene:
# sklearn.preprocessing  → preparación de datos (escalar, codificar...)
# sklearn.linear_model   → regresión lineal, logística, Lasso, Ridge...
# sklearn.ensemble       → Random Forest, Gradient Boosting...
# sklearn.svm            → Support Vector Machines
# sklearn.cluster        → K-Means, DBSCAN...
# sklearn.metrics        → métricas de evaluación (R², RMSE, accuracy...)
# sklearn.pipeline       → encadenar pasos de ML
from sklearn.preprocessing import StandardScaler
```

```python
# ───  Paleta de colores  ──────────────────────────────
# Colores semánticos usados en todas las gráficas del notebook
C_MAL   = "#D85A30"   # naranja/rojo  → indica problema o valor fuera de umbral
C_BIEN  = "#1D9E75"   # verde         → indica que todo está correcto
C_AZUL  = "#378ADD"   # azul          → líneas de referencia o tendencia
C_MORAD = "#7F77DD"   # morado        → elementos neutros / informativos
C_GRIS  = "#888780"   # gris          → cero o línea base


# ════════════════════════════════════════════════════════════════
#  FUNCIONES AUXILIARES
# ════════════════════════════════════════════════════════════════

def separador(n, titulo):
    """
    Imprime un separador visual en consola para organizar la salida
    del notebook en secciones numeradas.

    Ejemplo de salida:
    ═════════════════════════════════════════════════════════════════
      1. Normalidad de residuos
    ═════════════════════════════════════════════════════════════════
    """
    print(f"\n{'═'*65}")
    print(f"  {n}. {titulo}")
    print(f"{'═'*65}")


def graficar_qq(ax, residuos, color, titulo):
    """
    Dibuja un Q-Q plot usando statsmodels sobre un eje dado.
    Sirve para verificar visualmente si los residuos siguen una distribución normal.

    - Si los puntos siguen la línea azul → normalidad ✅
    - Si se desvían en los extremos     → no normalidad ❌

    Parámetros:
        ax       : eje de matplotlib donde se dibuja
        residuos : array con los errores del modelo
        color    : color de los puntos (C_BIEN o C_MAL según contexto)
        titulo   : texto del título del gráfico
    """


    # qqplot dibuja directamente sobre el eje 'ax'
    # line='s' → recta estandarizada (media y desviación de la muestra)
    sm.qqplot(residuos, line="s", ax=ax,
              markerfacecolor=color,
              markeredgecolor=color,
              alpha=0.65,
              markersize=4)

    # Personaliza la línea de referencia con el color del proyecto
    ax.get_lines()[0].set_color(C_AZUL)   # línea de referencia
    ax.set(xlabel="Cuantiles teóricos", ylabel="Cuantiles muestra",
           title=titulo)
    ax.grid(True, alpha=0.25)


def barra_vif(ax, vif_df, titulo):
    """
    Dibuja un gráfico de barras con los valores VIF (Variance Inflation Factor)
    de cada variable, para detectar multicolinealidad.

    Código de colores:
        - Verde (C_BIEN) → VIF ≤ 5, sin problema de multicolinealidad
        - Rojo  (C_MAL)  → VIF > 5, posible multicolinealidad

    Líneas de referencia:
        - Discontinua roja  → VIF = 5  (umbral moderado)
        - Punteada oscura   → VIF = 10 (umbral grave)

    Parámetros:
        ax     : eje de matplotlib donde se dibuja
        vif_df : DataFrame con columnas ["Variable", "VIF"]
        titulo : texto del título del gráfico
    """
    # Asigna color rojo si VIF > 5, verde si está dentro del umbral
    colores = [C_MAL if v > 5 else C_BIEN for v in vif_df["VIF"]]

    bars = ax.bar(vif_df["Variable"], vif_df["VIF"],
                  color=colores, edgecolor="white", width=0.45)

    # Líneas horizontales de referencia en VIF = 5 y VIF = 10
    ax.axhline(5,  color=C_MAL,  lw=1.2, ls="--", label="VIF = 5")
    ax.axhline(10, color="#8B0000", lw=1.2, ls=":",  label="VIF = 10")

    ax.set(ylabel="VIF", title=titulo)
    ax.legend(fontsize=8)

    # Ajusta el eje Y para que siempre haya margen visual suficiente
    ax.set_ylim(0, max(vif_df["VIF"].max() * 1.25, 12))

    # Añade el valor numérico encima de cada barra
    for bar, v in zip(bars, vif_df["VIF"]):
        ax.text(bar.get_x() + bar.get_width() / 2,
                bar.get_height() + 0.15, f"{v:.1f}",
                ha="center", va="bottom", fontsize=9, fontweight="bold")


def verificar(nombre_test, valor, umbral, sentido="mayor"):
    """
    Evalúa si un test estadístico supera (o no) un umbral,
    e imprime el resultado con un símbolo visual claro.

    Ejemplo de salida:
        ✓  Durbin-Watson: 1.9832  (umbral mayor 1.5)
        ✗  Shapiro-Wilk:  0.0210  (umbral mayor 0.05)

    Parámetros:
        nombre_test : nombre descriptivo del test (str)
        valor       : valor obtenido del test (float)
        umbral      : valor de referencia para comparar (float)
        sentido     : "mayor" → ok si valor > umbral  (por defecto)
                      "menor" → ok si valor < umbral

    Retorna:
        bool → True si el test pasa, False si no
    """
    if sentido == "mayor":
        ok = valor > umbral
    else:
        ok = valor < umbral

    marca = "✓" if ok else "✗"
    print(f"  {marca}  {nombre_test}: {valor:.4f}  (umbral {sentido} {umbral})")
    return ok
```

```python
# ════════════════════════════════════════════════════════════════
#  GENERADOR BASE DE DATOS (3 variables, dos correlacionadas)
# ════════════════════════════════════════════════════════════════

def generar_base(n, seed):
    # Se crea un generador de números aleatorios reproducible con la semilla dada
    # (misma semilla → mismos datos siempre)
    rng = np.random.default_rng(seed)

    # Se generan n superficies aleatorias en metros cuadrados, entre 40 y 200 m²
    x1 = rng.uniform(40, 200, n)

    # Se generan n valores de habitaciones derivados de la superficie:
    # dividimos x1/25 (aprox. 1 habitación cada 25m²) y añadimos ruido gaussiano
    # small (σ=0.4). Se redondea al entero más cercano y se limita entre 1 y 9
    # habitaciones. Esto hace que x1 y x2 estén correlacionadas → VIF alto
    x2 = np.round(x1 / 25 + rng.normal(0, 0.4, n)).clip(1, 9)

    # Se generan n distancias aleatorias al centro, entre 0.5 y 30 km
    # Es independiente de x1 y x2 → no genera multicolinealidad
    x3 = rng.uniform(0.5, 30, n)

    return x1, x2, x3


# ════════════════════════════════════════════════════════════════
#  1. LINEALIDAD
# ════════════════════════════════════════════════════════════════

separador(1, "LINEALIDAD — efecto cuadrático en distancia al centro")

# Tamaño del dataset para este experimento
n1 = 100

# Generador reproducible con semilla fija (seed=42)
rng1 = np.random.default_rng(42)

# Se generan 100 superficies entre 50 y 180 m²
x1 = rng1.uniform(50, 180, n1)

# Se generan habitaciones correlacionadas con la superficie (≈ 1 hab. cada 25m²)
# con ruido gaussiano σ=0.35, redondeadas y limitadas entre 1 y 8
x2 = np.round(x1 / 25 + rng1.normal(0, 0.35, n1)).clip(1, 8)

# Se generan 100 distancias al centro entre 1 y 25 km, independientes de x1 y x2
x3 = rng1.uniform(1, 25, n1)

# Se calcula el precio real con la siguiente estructura:
#   30.000      → precio base mínimo (intercepto)
#   +600·x1     → cada m² adicional sube el precio en 600€
#   +5.000·x2   → cada habitación adicional sube el precio en 5.000€
#   -4.500·x3   → alejarse del centro reduce el precio en 4.500€/km
#   +120·x3²    → efecto cuadrático: muy lejos vuelve a subir (zonas tranquilas)
#   +ruido      → variabilidad aleatoria del mercado (σ=6.000€)
precio1 = (
    30_000
    + 600   * x1
    + 5000  * x2
    - 4500  * x3
    + 120   * x3**2        # ← esta curvatura es lo que el modelo MALO no captura
    + rng1.normal(0, 6000, n1)
)

# Se empaqueta todo en un DataFrame con nombres de columna descriptivos
df1 = pd.DataFrame({"superficie": x1, "habitaciones": x2,
                    "distancia": x3,  "precio": precio1})

# Se muestran las primeras 8 filas para inspección visual rápida
df1.head(8)
```

```python
def ajustar_ols(X, y, titulo=""):
    """OLS con constante. Imprime resumen."""
    # OLS = Ordinary Least Squares (Mínimos Cuadrados Ordinarios)

    # Añade una columna de unos (β₀) a la matriz X
    # has_constant="add" la añade siempre aunque ya exista una columna constante
    Xc = sm.add_constant(X, has_constant="add")
    # sklearn añade el intercepto automáticamente:
    # LogisticRegression(fit_intercept=True)  # ← True por defecto

    # Ajusta el modelo OLS (mínimos cuadrados ordinarios) y lo entrena
    m  = sm.OLS(y, Xc).fit()

    # Si se proporciona título, imprime el encabezado con las métricas globales
    # R² → bondad del ajuste (cuánta varianza explica el modelo)
    # AIC → penaliza la complejidad; menor AIC = mejor modelo
    if titulo:
        print(f"\n  ── {titulo} ──")
        print(f"  R² = {m.rsquared:.4f}   AIC = {m.aic:.1f}")

    # Itera sobre cada variable del modelo imprimiendo sus coeficientes:
    #   v    → nombre de la variable
    #   b    → valor del coeficiente β (cuánto aporta al precio)
    #   p    → p-valor (probabilidad de que el coeficiente sea 0 por azar)
    #   star → "**" si p < 0.05 (estadísticamente significativo), "  " si no
    for v, b, p in zip(m.params.index, m.params, m.pvalues):
        star = "**" if p < 0.05 else "  "
        print(f"  {star} {v:22s}  β={b:10.4f}  p={p:.4f}")

    return m


def calcular_vif(df_X):
    """Devuelve DataFrame con VIF de cada columna."""

    # Añade la constante antes de calcular el VIF
    # (statsmodels la requiere para el cálculo correcto)
    Xc = sm.add_constant(df_X, has_constant="add")

    # Calcula el VIF para cada variable original (i+1 para saltar la constante)
    # variance_inflation_factor(matriz, índice_columna)
    return pd.DataFrame({
        "Variable": df_X.columns,
        "VIF": [variance_inflation_factor(Xc.values, i + 1)
                for i in range(df_X.shape[1])]  # i+1 → salta la columna constante
    })


# ── Uso ──────────────────────────────────────────────────────────

# Ajusta el modelo lineal con las 3 variables como predictores
# y el precio como variable objetivo
m1_inicial = ajustar_ols(
    df1[["superficie", "habitaciones", "distancia"]],
    df1["precio"],
    "MODELO INICIAL — lineal"
)

# Calcula el VIF de cada variable para detectar multicolinealidad
vif1_inicial = calcular_vif(df1[["superficie", "habitaciones", "distancia"]])

# Muestra la tabla VIF sin índice de fila para mayor legibilidad
print("\n  VIF modelo inicial:")
print(vif1_inicial.to_string(index=False))
```

```python
# ════════════════════════════════════════════════════════════════
#  2. INDEPENDENCIA DE ERRORES (Autocorrelación)
#  El error de hoy depende del error de ayer.
#  En una regresión ideal, cada error es independiente —
#  que el modelo se equivoque hoy no dice nada sobre si se equivocará mañana.
#  Cuando esto no ocurre, hay autocorrelación.
#  Lo veremos con una serie temporal de precios de alquiler diario
# ════════════════════════════════════════════════════════════════

separador(2, "AUTOCORRELACIÓN — alquiler diario con memoria AR(1)")

# 90 observaciones → simula 3 meses de datos diarios
n2 = 90
rng2 = np.random.default_rng(77)

# ── Predictores con componente temporal ──────────────────────────

# Índice de tiempo del día 1 al 90
t = np.arange(1, n2 + 1)

# Superficie que crece con el tiempo (mercado en expansión)
# +2m² por día de media, con ruido gaussiano σ=8
x1_t = 40 + 2.0 * t + rng2.normal(0, 8, n2)

# Habitaciones derivadas de la superficie, igual que antes
x2_t = np.round(x1_t / 25 + rng2.normal(0, 0.4, n2)).clip(1, 9)

# Distancia al centro, aleatoria e independiente del tiempo
x3_t = rng2.uniform(0.5, 20, n2)

# ── Error AR(1) ───────────────────────────────────────────────────
# AR(1) = AutoRegresivo de orden 1:
# el error de hoy depende del error de ayer
# ε[t] = ρ·ε[t-1] + ruido
#
# ρ = 0.80 → correlación alta entre errores consecutivos
# (viola el supuesto de independencia de errores)
rho = 0.80
eps = np.zeros(n2)                          # inicializa errores a 0
for i in range(1, n2):
    # Cada error hereda el 80% del error anterior más ruido nuevo
    eps[i] = rho * eps[i - 1] + rng2.normal(0, 50)

# Precio real generado con la estructura lineal + error autocorrelacionado
# 200      → precio base
# 5·x1_t   → efecto superficie
# 120·x2_t → efecto habitaciones
# -15·x3_t → efecto distancia
# + eps    → error con memoria (el problema que queremos detectar)
precio2 = 200 + 5 * x1_t + 120 * x2_t - 15 * x3_t + eps

# Empaqueta todo en un DataFrame
df2 = pd.DataFrame({
    "t": t, "superficie": x1_t, "habitaciones": x2_t,
    "distancia": x3_t, "precio": precio2
})

# Añade el precio del día anterior como variable adicional (lag 1)
# shift(1) → desplaza una posición hacia abajo
# bfill()  → rellena el NaN inicial con el valor siguiente
df2["precio_lag1"] = df2["precio"].shift(1).bfill()

# ── Modelo inicial (incorrecto) ───────────────────────────────────

# Ajusta el modelo lineal estándar ignorando la autocorrelación
# Este modelo violará el supuesto de independencia de errores
m2_inicial = ajustar_ols(
    df2[["superficie", "habitaciones", "distancia"]],
    df2["precio"],
    "MODELO INCORRECTO"
)

# ── Figura diagnóstica ────────────────────────────────────────────

fig, axes = plt.subplots(1, 2, figsize=(16, 4))
fig.suptitle("2. INDEPENDENCIA DE ERRORES (AUTOCORRELACIÓN)",
             fontweight="bold", fontsize=13)

# Panel 0: serie temporal real vs ajuste del modelo
# Si el modelo fuera perfecto, ambas líneas coincidirían
axes[0].plot(t, df2["precio"], color=C_AZUL, lw=1.2, label="precio real")
axes[0].plot(t, m2_inicial.fittedvalues, color=C_MAL,
             lw=1.4, ls="--", label="ajuste")
axes[0].set(xlabel="Día", ylabel="Precio (€)",
            title="Serie temporal de precios")
axes[0].legend(fontsize=8)

# Panel 1: residuos a lo largo del tiempo
# Si hay autocorrelación los residuos formarán ondas o rachas
# (no serán aleatorios alrededor del cero)
axes[1].plot(t, m2_inicial.resid, color=C_MAL, lw=1.2)
axes[1].axhline(0, color=C_GRIS, lw=1.2, ls="--")
axes[1].set(xlabel="Día", ylabel="Residuo",
            title="Residuos en el tiempo — ANTES")
axes[1].grid(alpha=0.2)

# Imprime el resumen estadístico completo del modelo
# Incluye R², coeficientes, p-valores y el estadístico Durbin-Watson
print(m2_inicial.summary())

plt.tight_layout()
plt.savefig("2_autocorrelacion.png", dpi=150, bbox_inches="tight")
plt.show()
print("  → Guardado: 2_autocorrelacion.png")
```

```python
# Añade el precio del día anterior como nueva variable predictora
# shift(1) → desplaza el precio un día hacia adelante
#            (el precio de hoy pasa a ser el precio de mañana)
# bfill()  → el primer valor sería NaN (no hay día anterior),
#            lo rellena con el siguiente valor disponible
df2["precio_lag1"] = df2["precio"].shift(1).bfill()

# ── Modelo corregido ──────────────────────────────────────────────
# Añade precio_lag1 como predictor para que el modelo pueda
# "aprender" la memoria temporal de los errores:
# si ayer el precio fue alto, hoy probablemente también lo sea
m2_bien = ajustar_ols(
    df2[["superficie", "habitaciones", "distancia", "precio_lag1"]],
    df2["precio"],
    "MODELO CORREGIDO — con lag-1"
)

# Muestra el resumen completo del modelo corregido
# Nos interesa especialmente el Durbin-Watson del summary
print(m2_bien.summary())

# ── Diagnóstico numérico: Durbin-Watson ───────────────────────────

# Calcula el estadístico Durbin-Watson para ambos modelos
# DW ≈ 0 → autocorrelación positiva fuerte ❌
# DW ≈ 2 → sin autocorrelación            ✅
# DW ≈ 4 → autocorrelación negativa       ❌
dw_inicial = sm.stats.durbin_watson(m2_inicial.resid)
dw_bien    = sm.stats.durbin_watson(m2_bien.resid)

# ── Comparación antes / después ───────────────────────────────────
print("\n── Verificación ──")

# DW del modelo malo → esperamos valor cercano a 0 (autocorrelación alta)
verificar("Durbin-Watson (ideal ≈ 2)", dw_inicial, 1.5, "mayor")

# DW del modelo corregido → esperamos valor cercano a 2 (sin autocorrelación)
verificar("Durbin-Watson corregido",   dw_bien,    1.5, "mayor")
```

```python
# ════════════════════════════════════════════════════════════════
#  3. HOMOCEDASTICIDAD
#     Varianza proporcional a superficie (casas grandes, mayor ruido)
# ════════════════════════════════════════════════════════════════

# Crea figura con 4 paneles para diagnosticar la heterocedasticidad
# contra cada variable del modelo y contra los valores predichos
fig, axes = plt.subplots(1, 4, figsize=(22, 4))
fig.suptitle("3. Heterocedasticidad", fontweight="bold", fontsize=18, y=1.2)


# ── Panel 0: Residuos vs Precio predicho ─────────────────────────
# Si la varianza de los residuos crece con el precio predicho
# → heterocedasticidad (forma de embudo o abanico)
axes[0].scatter(m2_bien.fittedvalues, m2_bien.resid,
                color=C_MAL, alpha=0.55, s=22)
axes[0].set(xlabel="Precio (€)", ylabel="Residuos",
            title="Residuos  vs Precio (€)")
axes[0].axhline(0, color=C_GRIS, lw=1.2, ls="--")
axes[0].grid(alpha=0.2)

# ── Panel 1: Residuos vs Superficie ──────────────────────────────
# Panel clave para este supuesto: si casas más grandes tienen
# errores más grandes → varianza crece con la superficie → ❌
axes[1].scatter(df2["superficie"], m2_bien.resid,
                color=C_MAL, alpha=0.55, s=22)
axes[1].set(xlabel="Superficie", ylabel="",
            title="Residuos  vs Superficie")
axes[1].axhline(0, color=C_GRIS, lw=1.2, ls="--")
axes[1].grid(alpha=0.2)

# ── Panel 2: Residuos vs Habitaciones ────────────────────────────
# Al estar correlacionada con superficie, puede mostrar
# un patrón similar de varianza creciente
axes[2].scatter(df2["habitaciones"], m2_bien.resid,
                color=C_MAL, alpha=0.55, s=22)
axes[2].set(xlabel="Habitaciones", ylabel="",
            title="Residuos  vs Habitaciones")
axes[2].axhline(0, color=C_GRIS, lw=1.2, ls="--")
axes[2].grid(alpha=0.2)

# ── Panel 3: Residuos vs Distancia ───────────────────────────────
# Variable independiente de la superficie → no debería
# mostrar patrón de varianza creciente si el problema
# está solo en la superficie
axes[3].scatter(df2["distancia"], m2_bien.resid,
                color=C_MAL, alpha=0.55, s=22)
axes[3].set(xlabel="Distancia", ylabel="",
            title="Residuos  vs Distancia")
axes[3].axhline(0, color=C_GRIS, lw=1.2, ls="--")
axes[3].grid(alpha=0.2)
```

```python
# ── Diagnóstico numérico: Test de Breusch-Pagan ──────────────────
# Contrasta formalmente si la varianza de los residuos es constante
# H₀: homocedasticidad (varianza constante)     → p >= 0.05 ✅
# H₁: heterocedasticidad (varianza no constante) → p <  0.05 ❌
#
# Necesita:
#   - los residuos del modelo
#   - la matriz de predictores (con constante añadida)
#
# Devuelve 4 valores, pero solo usamos los 2 primeros:
#   bp_stat → estadístico del test (no se usa aquí)
#   bp_p    → p-valor (el que nos interesa)
#   _       → f-statistic (descartado)
#   _       → f p-valor   (descartado)
bp_stat, bp_p, _, _ = sm.stats.het_breuschpagan(
    m2_bien.resid,
    sm.add_constant(df2[["superficie", "habitaciones", "distancia", "precio_lag1"]])
)

# ── Corrección: transformación logarítmica ────────────────────────
# Si existe heterocedasticidad, una solución habitual es
# modelar log(precio) en lugar de precio directamente.
# El logaritmo comprime los valores grandes → estabiliza la varianza
df2["log_precio"] = np.log(df2["precio"])

# Ajusta el modelo corregido con log(precio) como variable objetivo
# Usa los mismos predictores que el modelo anterior
m3_inicial = ajustar_ols(
    df2[["superficie", "habitaciones", "distancia", "precio_lag1"]],
    df2["log_precio"],
    "MODELO CORREGIDO — log(precio)"
)

# ── Verificación tras la corrección ──────────────────────────────
# Aplica el test de Breusch-Pagan al modelo corregido
# para comprobar si el logaritmo resolvió el problema
bp_s2, bp_p2, _, _ = sm.stats.het_breuschpagan(
    m3_inicial.resid,
    sm.add_constant(df2[["superficie", "habitaciones", "distancia", "precio_lag1"]])
)

# ── Imprime la comparación antes / después ────────────────────────
print("\n── Verificación ──")

# p < 0.05 → rechazamos H₀ → hay heterocedasticidad ❌
# p >= 0.05 → no rechazamos H₀ → homocedasticidad ✅
print(f"  Breusch-Pagan p ANTES   = {bp_p:.4f}  {'⚠ Heteroced.' if bp_p < 0.05  else '✓'}")
print(f"  Breusch-Pagan p DESPUÉS = {bp_p2:.4f}  {'✓ Homoced.'   if bp_p2 >= 0.05 else '⚠'}")
```

```python
#  Otras transformaciones de Y
# Raíz cuadrada → compresión más suave que el log
df2["sqrt_precio"] = np.sqrt(df2["precio"])

# Inversa → compresión más agresiva
df2["inv_precio"] = 1 / df2["precio"]

# Box-Cox → encuentra automáticamente la mejor transformación
from scipy.stats import boxcox
df2["bc_precio"], lambda_optimo = boxcox(df2["precio"])



# Ajusta el modelo corregido con otras transformaciones como variable objetivo
# Usa los mismos predictores que el modelo anterior
m4 = ajustar_ols(
    df2[["superficie", "habitaciones", "distancia"]],
    df2["sqrt_precio"],
    "MODELO CORREGIDO — sqrt_precio"
)

# ── Verificación tras la corrección ──────────────────────────────
# Aplica el test de Breusch-Pagan al modelo corregido
# para comprobar si el logaritmo resolvió el problema
bp_s2, bp_p2, _, _ = sm.stats.het_breuschpagan(
    m4.resid,
    sm.add_constant(df2[["superficie", "habitaciones", "distancia"]])
)

# ── Imprime la comparación antes / después ────────────────────────
print("\n── Verificación ──")

# p < 0.05 → rechazamos H₀ → hay heterocedasticidad ❌
# p >= 0.05 → no rechazamos H₀ → homocedasticidad ✅
print(f"  Breusch-Pagan p ANTES   = {bp_p:.4f}  {'⚠ Heteroced.' if bp_p < 0.05  else '✓'}")
print(f"  Breusch-Pagan p DESPUÉS = {bp_p2:.4f}  {'✓ Homoced.'   if bp_p2 >= 0.05 else '⚠'}")
```

```python
# Ahora, ajustamos el mismo modelo anterior pero con las variables estandarizadas.
# El objetivo es poder comparar la importancia relativa de cada predictor.


# 1. Define predictores y variable objetivo (igual que antes)
X = df2[["superficie", "distancia", "precio_lag1"]]
y = df2["precio"]

# 2. Crea el escalador y transforma X
scaler   = StandardScaler()
X_scaled = scaler.fit_transform(X)
# fit_transform hace dos cosas:
#   fit()       → calcula media (μ) y desviación típica (σ) de cada variable
#   transform() → aplica la fórmula: X_scaled = (X - μ) / σ
#
# Resultado: cada variable tiene media=0 y desviación típica=1


# 3. Ajusta OLS con las variables escaladas
m_final_escalado = ajustar_ols(X_scaled, df2["precio"], "MODELO ESCALADO")
print(m_final_escalado.summary())
```

## 🚨 Los Coeficientes hallados, están escalados, no son los reales de la regresión.

$$
y = \beta_0 + \beta_1 z = \beta_0 + \beta_1 \left( \frac{X - \mu}{\sigma} \right) = \underbrace{\left( \beta_0 - \frac{\beta_1 \mu}{\sigma} \right)}_{\text{Nuevo intercepto}} + \underbrace{\left( \frac{\beta_1}{\sigma} \right)}_{\text{Nueva pendiente}} X
$$

```python
# Recupera la media de cada variable aprendida por el scaler
scaler_means = scaler.mean_
# [μ_superficie, μ_distancia, μ_precio_lag1]

# Recupera la desviación típica de cada variable
scaler_stds = np.sqrt(scaler.var_)
# [σ_superficie, σ_distancia, σ_precio_lag1]

# ── Intercepto original ───────────────────────────────────────────
# La fórmula ajusta el intercepto para compensar el desplazamiento
# introducido por restar las medias durante el escalado
intercepto_original = (m_final_escalado.params.iloc[0]              # β₀ escalado
                       - np.sum(m_final_escalado.params.iloc[1:]     # Σ βᵢ escalados
                       * scaler_means / scaler_stds))       # × (μᵢ/σᵢ)

# ── Coeficientes originales ───────────────────────────────────────
# Para cada β escalado, dividir por σ de su variable
# βᵢ_original = βᵢ_escalado / σᵢ
coef_metros_original = m_final_escalado.params.iloc[1] / scaler_stds[0]  # β_superficie
coef_planta_original = m_final_escalado.params.iloc[2] / scaler_stds[1]  # β_distancia
coef_dist_original   = m_final_escalado.params.iloc[3] / scaler_stds[2]  # β_precio_lag1
```

```python
print("Intercepto_original:", intercepto_original)
print("superficie:",coef_metros_original)
print("distancia:",coef_planta_original)
print("precio_lag1:  ",coef_dist_original)
```
