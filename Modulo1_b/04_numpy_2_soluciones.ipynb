```python
# @title
%%html

<link rel="stylesheet" href="https://cdn.jsdelivr.net/npm/@tabler/icons-webfont@latest/tabler-icons.min.css">
<style>
  @import url('https://fonts.googleapis.com/css2?family=Source+Code+Pro:wght@400;500&family=DM+Sans:wght@300;400;500;600&display=swap');
  .nb-cell{font-family:'DM Sans',sans-serif;padding:1.2rem 0;max-width:860px;}
  .nb-divider{height:2px;background:linear-gradient(90deg,#1E3A8A 0%,#475569 60%,transparent 100%);margin-bottom:1.1rem;border:none;}
  code.ic{font-family:'Source Code Pro',monospace;font-size:12.5px;background:#EFF6FF;color:#1E3A8A;padding:2px 7px;border-radius:4px;border:1px solid #BFDBFE;white-space:nowrap;}
  code.ic-g{font-family:'Source Code Pro',monospace;font-size:12.5px;background:#F0FDF4;color:#0F766E;padding:2px 7px;border-radius:4px;border:1px solid #99F6E4;white-space:nowrap;}
  .nb-note{display:flex;gap:10px;align-items:flex-start;background:#F0FDF4;border-left:3px solid #0F766E;border-radius:0 6px 6px 0;padding:10px 14px;margin-top:.8rem;}
  .nb-note-text{font-size:13px;color:#0f766e;line-height:1.6;}
  .hint-box{display:flex;align-items:flex-start;gap:10px;background:#FFF7ED;border-left:3px solid #F59E0B;border-radius:0 6px 6px 0;padding:10px 14px;margin-top:.8rem;}
  .hint-text{font-size:13px;color:#92400E;line-height:1.6;}
  .ex-card{border-radius:10px;overflow:hidden;border:1px solid rgba(0,0,0,.08);box-shadow:0 2px 8px rgba(0,0,0,.05);}
  .ex-hdr{display:flex;align-items:center;gap:12px;padding:10px 16px;background:#1E3A8A;}
  .ex-num{display:flex;align-items:center;justify-content:center;width:26px;height:26px;border-radius:50%;background:rgba(255,255,255,.15);color:#fff;font-size:13px;font-weight:700;flex-shrink:0;}
  .ex-title{font-size:13px;font-weight:600;color:rgba(255,255,255,.95);letter-spacing:.2px;}
  .ex-body{background:#F8FAFF;padding:12px 16px;}
  .ex-task{font-size:13.5px;color:#1e293b;line-height:1.65;margin:0 0 .5rem 0;}
  .ex-task:last-child{margin-bottom:0;}
</style>

<div class="nb-cell">
  <div class="ex-card">
    <div class="ex-hdr">
      <div class="ex-num">4</div>
      <div class="ex-title">Matriz de tablero (índices pares → 1)</div>
    </div>
    <div class="ex-body"><p class="ex-task">▸ Obtén una matriz (7×7) de ceros donde la suma de sus índices sea par → convertir en 1.</p><p class="ex-task">→ Ejecuta la celda de visualización para ver el resultado como imagen.</p></div>
  </div>
  <div class="hint-box"><i class="ti ti-bulb" style="color:#D97706;font-size:15px;flex-shrink:0;margin-top:1px;"></i><span class="hint-text">Recorre la matriz con dos bucles <code class='ic'>for</code> y comprueba <code class='ic'>(i+j) % 2 == 0</code>.</span></div><div class="nb-note"><i class="ti ti-info-circle" style="color:#0F766E;font-size:15px;flex-shrink:0;margin-top:1px;"></i><span class="nb-note-text">El resultado es un patrón de tablero de ajedrez. La visualización con <code class='ic'>imshow(cmap='gray')</code> lo confirma.</span></div>
</div>
```

```python
import numpy as np
import matplotlib.pyplot as plt

matriz = np.zeros((7, 7))

# Si la suma de los índices (i+j) es par → 1
for i in range(7):
    for j in range(7):
        if (i + j) % 2 == 0:
            matriz[i, j] = 1

print(matriz)

# Visualización en escala de grises (tablero de ajedrez)
plt.figure(figsize=(5, 5))
plt.imshow(matriz, cmap="gray")
plt.title("Patrón par/impar")
plt.show()
```

```python
import matplotlib.pyplot as plt

def tablero_patron(n):
    """Genera una matriz nxn con patrón de tablero de ajedrez."""
    M = np.zeros((n, n), dtype=int)
    M[::2, ::2]   = 1
    M[1::2, 1::2] = 1
    return M

# Mostrar para varios tamaños
fig, axes = plt.subplots(1, 4, figsize=(14, 4))
for ax, n in zip(axes, [4, 8, 16, 32]):
    ax.imshow(tablero_patron(n), cmap="gray", interpolation="nearest")
    ax.set_title(f"{n}×{n}")
    ax.axis("off")
plt.suptitle("Patrón de tablero — distintos tamaños", fontsize=13)
plt.tight_layout()
plt.show()
```
