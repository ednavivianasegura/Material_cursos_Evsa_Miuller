```python
pares   = [i for i in range(1, 20) if i % 2 == 0][:5]   # [2, 4, 6, 8, 10]
impares = [i for i in range(1, 20) if i % 2 != 0][:5]   # [1, 3, 5, 7, 9]

# column_stack combina los vectores como columnas → (5, 2)
matriz = np.column_stack((pares, impares))
print(matriz)
print("shape:", matriz.shape)
```

```python
def primeros_primos(n):
    """Devuelve los primeros n números primos."""
    primos, candidato = [], 2
    while len(primos) < n:
        if all(candidato % p != 0 for p in primos):
            primos.append(candidato)
        candidato += 1
    return primos

# Rellenar la matriz vacía (3x6 = 18 elementos)
M_primos = np.empty((3, 6), dtype="float64")
M_primos[:] = np.array(primeros_primos(18)).reshape(3, 6)
print(M_primos)
```

```python
np.random.seed(42)   # semilla para reproducibilidad
A = np.random.randint(1, 10, size=(4, 4))
B = np.random.randint(1, 10, size=(4, 4))
print("A:\n", A)
print("B:\n", B)

# Opción A: concatenate con axis=0
resultado = np.concatenate((A, B), axis=0)
print("\nconcatenate axis=0 (8x4):\n", resultado)

# Opción B: vstack (equivalente)
resultado2 = np.vstack((A, B))
print("\nvstack (8x4):\n", resultado2)
print("shape:", resultado.shape)
```
