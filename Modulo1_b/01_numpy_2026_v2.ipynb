```python
vector = np.array([-2, 4, 1, 3.5])

print("argmin:", vector.argmin())
print("argmax:", vector.argmax())
print("cumsum:", vector.cumsum())
print("cumprod:", vector.cumprod())

matriz_1 = vector.reshape(2, 2)
print("reshape(2,2):\n", matriz_1)
print("transpose:\n", matriz_1.T)
print("flatten:", matriz_1.flatten())
print("tolist:", matriz_1.tolist())

print("\nAntes de sort:", vector)
vector.sort()
print("Después de sort:", vector)
```

```python
# @title
%%html

<link rel="stylesheet" href="https://cdn.jsdelivr.net/npm/@tabler/icons-webfont@latest/tabler-icons.min.css">
<style>
  @import url('https://fonts.googleapis.com/css2?family=Source+Code+Pro:wght@400;500&family=DM+Sans:wght@300;400;500;600&display=swap');
  .nb-cell{font-family:'DM Sans',sans-serif;padding:1.5rem 0;max-width:860px;}
  .nb-title{font-size:20px;font-weight:600;color:#1E3A8A;margin:0 0 1rem 0;letter-spacing:-.3px;}
  .nb-title-lg{font-size:22px;font-weight:600;color:#1E3A8A;margin:0;letter-spacing:-.3px;}
  .nb-badge{font-size:11px;font-weight:500;color:#0F766E;background:#CCFBF1;padding:3px 10px;border-radius:20px;border:1px solid #99F6E4;letter-spacing:.4px;text-transform:uppercase;}
  .nb-divider{height:2px;background:linear-gradient(90deg,#1E3A8A 0%,#475569 60%,transparent 100%);margin-bottom:1.25rem;border:none;}
  .nb-subtitle{font-size:13px;font-weight:500;color:#475569;text-transform:uppercase;letter-spacing:.8px;margin-bottom:1rem;}
  .nb-body{font-size:14px;color:#1e293b;line-height:1.75;margin-bottom:1.25rem;}
  .nb-body p{margin:0 0 .75rem;}
  .nb-body-box{font-size:14px;color:#1e293b;line-height:1.75;margin-bottom:1.25rem;padding:10px 14px;background:#F8FAFF;border-left:3px solid #1E3A8A;border-radius:0 6px 6px 0;}
  /* code */
  code.ic{font-family:'Source Code Pro',monospace;font-size:13px;background:#EFF6FF;color:#1E3A8A;padding:2px 8px;border-radius:4px;border:1px solid #BFDBFE;cursor:help;white-space:nowrap;}
  code.ic-lg{font-family:'Source Code Pro',monospace;font-size:13.5px;background:#EFF6FF;color:#1E3A8A;padding:1px 7px;border-radius:4px;border:1px solid #BFDBFE;}
  code.at{font-family:'Source Code Pro',monospace;font-size:13px;background:#F0FDF4;color:#0F766E;padding:2px 8px;border-radius:4px;border:1px solid #99F6E4;cursor:help;}
  code.pm{font-family:'Source Code Pro',monospace;font-size:12px;background:#FFF7ED;color:#92400E;padding:1px 6px;border-radius:4px;border:1px solid #FED7AA;}
  /* dark code block */
  .code-block{background:#0f172a;border-radius:8px;overflow:hidden;margin-top:.5rem;}
  .code-header{display:flex;align-items:center;justify-content:space-between;padding:8px 16px;background:#1e293b;border-bottom:1px solid #334155;}
  .code-lang{font-family:'Source Code Pro',monospace;font-size:11px;color:#64748b;letter-spacing:.5px;text-transform:uppercase;}
  .dots{display:flex;gap:5px;}.dot{width:10px;height:10px;border-radius:50%;}
  .dot-r{background:#ef4444;}.dot-y{background:#f59e0b;}.dot-g{background:#22c55e;}
  pre.cc{padding:16px 20px;font-family:'Source Code Pro',monospace;font-size:14px;line-height:1.7;color:#e2e8f0;margin:0;white-space:pre;}
  .kw{color:#7dd3fc;}.mod{color:#a5f3fc;}.al{color:#86efac;}.cm{color:#94A3B8;font-style:italic;}
  .arr{color:#86efac;}.idx{color:#7dd3fc;}.sep{color:#94A3B8;}.slc{color:#f9a8d4;}.brk{color:#fcd34d;}
  .op{color:#f9a8d4;}.num{color:#fcd34d;}.fn{color:#c4b5fd;}
  /* note */
  .nb-note{display:flex;gap:10px;align-items:flex-start;background:#F0FDF4;border-left:3px solid #0F766E;border-radius:0 6px 6px 0;padding:10px 14px;margin-top:1rem;}
  .nb-note-text{font-size:13.5px;color:#0f766e;line-height:1.6;}
  .warn-box{display:flex;align-items:flex-start;gap:12px;background:#FFF7ED;border-left:3px solid #F59E0B;border-radius:0 8px 8px 0;padding:14px 18px;}
  .warn-label{font-size:11px;font-weight:600;letter-spacing:1px;text-transform:uppercase;color:#B45309;margin-bottom:6px;}
  .warn-text{font-size:14px;color:#78350F;line-height:1.7;margin:0;}
  /* table */
  .table-wrap{border-radius:8px;overflow:hidden;border:1px solid rgba(0,0,0,.08);box-shadow:0 2px 8px rgba(0,0,0,.05);}
  .ref-table{width:100%;border-collapse:collapse;font-size:14px;}
  .ref-table thead tr{background:#1E3A8A;color:#fff;}
  .ref-table thead th{padding:12px 20px;text-align:left;font-weight:500;font-size:12px;letter-spacing:.8px;text-transform:uppercase;}
  .ref-table thead th:first-child{border-radius:8px 0 0 0;}
  .ref-table thead th:last-child{border-radius:0 8px 0 0;}
  .ref-table tbody tr:nth-child(odd){background:#F8FAFF;}
  .ref-table tbody tr:nth-child(even){background:#fff;}
  .ref-table tbody tr:hover{background:#EFF6FF;}
  .ref-table tbody td{padding:10px 20px;border-bottom:.5px solid rgba(0,0,0,.06);color:#1e293b;vertical-align:middle;}
  .ref-table tbody tr:last-child td{border-bottom:none;}
  .ref-table tbody tr:last-child td:first-child{border-radius:0 0 0 8px;}
  .ref-table tbody tr:last-child td:last-child{border-radius:0 0 8px 0;}
  .desc{font-size:13px;color:#475569;line-height:1.5;}
  .desc strong{color:#1E3A8A;font-weight:600;}
  /* tooltip */
  .tip{position:relative;display:inline-block;}
  .tip .tt{visibility:hidden;opacity:0;background:#1e293b;color:#e2e8f0;font-family:'DM Sans',sans-serif;font-size:12px;line-height:1.5;border-radius:6px;padding:7px 11px;position:absolute;z-index:999;left:0;top:calc(100% + 6px);white-space:normal;max-width:300px;box-shadow:0 4px 12px rgba(0,0,0,.2);transition:opacity .18s ease;pointer-events:none;}
  .tip .tt::before{content:'';position:absolute;bottom:100%;left:14px;border:5px solid transparent;border-bottom-color:#1e293b;}
  .tip:hover .tt{visibility:visible;opacity:1;}
  /* badges */
  .rb{display:inline-block;font-size:10px;font-weight:600;padding:2px 7px;border-radius:20px;vertical-align:middle;white-space:nowrap;}
  .rb-blue{background:#EFF6FF;color:#1E3A8A;border:1px solid #BFDBFE;}
  .rb-teal{background:#F0FDF4;color:#0F766E;border:1px solid #99F6E4;}
  .rb-amber{background:#FFF7ED;color:#92400E;border:1px solid #FED7AA;}
  .rb-purple{background:#F5F3FF;color:#5B21B6;border:1px solid #DDD6FE;}
  .rb-slate{background:#F8FAFF;color:#475569;border:1px solid #CBD5E1;}
  /* legend / hint */
  .legend{display:flex;gap:20px;margin-top:.9rem;flex-wrap:wrap;}
  .legend-item{display:flex;align-items:center;gap:6px;font-size:12px;color:#475569;}
  .hint{font-size:11px;color:#94A3B8;margin-top:.6rem;display:flex;align-items:center;gap:5px;}
  /* cards */
  .card{border-radius:10px;overflow:hidden;border:1px solid rgba(0,0,0,.08);box-shadow:0 2px 6px rgba(0,0,0,.04);}
  .card-hdr{display:flex;align-items:center;justify-content:space-between;padding:9px 14px;}
  .card-hdr.blue{background:#1E3A8A;}.card-hdr.teal{background:#0F766E;}.card-hdr.slate{background:#475569;}
  .card-label{font-size:11px;font-weight:600;letter-spacing:.8px;text-transform:uppercase;color:rgba(255,255,255,.9);}
  .card-badge{font-size:10px;background:rgba(255,255,255,.15);color:#fff;padding:2px 8px;border-radius:20px;}
  .sdots{display:flex;gap:4px;}.sdot{width:9px;height:9px;border-radius:50%;}
  .card-body{background:#0f172a;padding:14px 18px;}
  .card-desc{background:#F8FAFF;padding:10px 14px;font-size:13px;color:#475569;line-height:1.6;}
  .card-desc strong{color:#1E3A8A;}
  pre.sc{font-family:'Source Code Pro',monospace;font-size:13.5px;line-height:1.7;color:#e2e8f0;margin:0;white-space:pre;}
  /* op grid */
  .op-grid{display:grid;grid-template-columns:1fr 1fr;gap:12px;margin-bottom:12px;}
  .op-card{border-radius:10px;overflow:hidden;border:1px solid rgba(0,0,0,.08);box-shadow:0 2px 6px rgba(0,0,0,.04);}
  .op-hdr{display:flex;align-items:center;justify-content:space-between;padding:8px 14px;background:#475569;}
  .op-hdr.teal{background:#0F766E;}.op-hdr.blue{background:#1E3A8A;}
  .op-label{font-size:11px;font-weight:600;letter-spacing:.8px;text-transform:uppercase;color:rgba(255,255,255,.9);}
  .op-body{background:#0f172a;padding:12px 16px;}
  .op-desc{background:#F8FAFF;padding:9px 14px;font-size:12.5px;color:#475569;line-height:1.5;}
  .op-desc strong{color:#1E3A8A;}
  pre.oc{font-family:'Source Code Pro',monospace;font-size:13px;line-height:1.6;color:#e2e8f0;margin:0;white-space:pre;}
  /* concat */
  .concat-grid{display:grid;grid-template-columns:1fr 1fr;gap:12px;}
  .concat-card{border-radius:10px;overflow:hidden;border:1px solid rgba(0,0,0,.08);}
  .concat-hdr{padding:9px 14px;}.concat-hdr.v{background:#1E3A8A;}.concat-hdr.h{background:#0F766E;}
  .concat-label{font-size:11px;font-weight:600;letter-spacing:.8px;text-transform:uppercase;color:rgba(255,255,255,.9);}
  .concat-body{background:#0f172a;padding:12px 16px;}
  .concat-desc{background:#F8FAFF;padding:9px 14px;font-size:12.5px;color:#475569;line-height:1.5;}
  .concat-desc strong{color:#1E3A8A;}
  /* mini inline label */
  .inline-label{display:inline-block;font-size:10px;font-weight:600;background:#1e293b;color:#94A3B8;padding:1px 6px;border-radius:3px;font-family:'Source Code Pro',monospace;margin-right:4px;}
</style>

<div class="nb-cell">
  <h3 class="nb-title">Operaciones con Arrays</h3>
  <hr class="nb-divider">
  <div class="nb-body-box">
    NumPy aplica todas las operaciones <strong>elemento a elemento</strong> (<em>element-wise</em>),
    lo que significa que opera sobre cada par de elementos en la misma posición.
    Para el producto matricial matemático se usa <code class="ic-lg">.dot()</code> o el operador <code class="ic-lg">@</code>.
    Las matrices A y B usadas en los ejemplos son de forma <strong>2×2</strong>.
  </div>

  <div class="op-grid">

    <div class="op-card">
      <div class="op-hdr blue">
        <div class="sdots"><div class="sdot" style="background:#ef4444"></div><div class="sdot" style="background:#f59e0b"></div><div class="sdot" style="background:#22c55e"></div></div>
        <span class="op-label">Suma y Resta</span>
        <span class="card-badge">Element-wise</span>
      </div>
      <div class="op-body">
        <pre class="oc"><span class="al">A</span> <span class="op">+</span> <span class="al">B</span>   <span class="cm"># suma elemento a elemento</span>
<span class="al">A</span> <span class="op">-</span> <span class="al">B</span>   <span class="cm"># resta elemento a elemento</span></pre>
      </div>
      <div class="op-desc">Operan sobre cada par en la <strong>misma posición</strong>. Requieren que A y B tengan la misma forma.</div>
    </div>

    <div class="op-card">
      <div class="op-hdr slate">
        <div class="sdots"><div class="sdot" style="background:#ef4444"></div><div class="sdot" style="background:#f59e0b"></div><div class="sdot" style="background:#22c55e"></div></div>
        <span class="op-label">Multiplicación por escalar</span>
        <span class="card-badge">Broadcasting</span>
      </div>
      <div class="op-body">
        <pre class="oc"><span class="al">A</span> <span class="op">*</span> <span class="num">5</span>   <span class="cm"># multiplica cada elemento por 5</span>
<span class="al">A</span> <span class="op">/</span> <span class="num">5</span>   <span class="cm"># divide cada elemento entre 5</span></pre>
      </div>
      <div class="op-desc">El escalar se <strong>"transmite"</strong> (broadcasting) a todos los elementos del array.</div>
    </div>

    <div class="op-card">
      <div class="op-hdr slate">
        <div class="sdots"><div class="sdot" style="background:#ef4444"></div><div class="sdot" style="background:#f59e0b"></div><div class="sdot" style="background:#22c55e"></div></div>
        <span class="op-label">Multiplicación element-wise</span>
        <span class="card-badge">No es producto matricial</span>
      </div>
      <div class="op-body">
        <pre class="oc"><span class="al">A</span> <span class="op">*</span> <span class="al">B</span>   <span class="cm"># multiplica posición a posición</span></pre>
      </div>
      <div class="op-desc"><strong>No</strong> es el producto de matrices de álgebra lineal. Cada elemento de A se multiplica por el correspondiente de B.</div>
    </div>

    <div class="op-card">
      <div class="op-hdr teal">
        <div class="sdots"><div class="sdot" style="background:#ef4444"></div><div class="sdot" style="background:#f59e0b"></div><div class="sdot" style="background:#22c55e"></div></div>
        <span class="op-label">Producto matricial</span>
        <span class="card-badge">Álgebra lineal</span>
      </div>
      <div class="op-body">
        <pre class="oc"><span class="al">A</span>.dot(<span class="al">B</span>)   <span class="cm"># producto matricial</span>
<span class="al">A</span> <span class="op">@</span> <span class="al">B</span>      <span class="cm"># equivalente (Python 3.5+)</span></pre>
      </div>
      <div class="op-desc">Producto matricial estándar: combina filas de A con <strong>columnas de B</strong>. Requiere que el nº de columnas de A = nº de filas de B.</div>
    </div>

  </div>

  <div class="hint">
    <i class="ti ti-info-circle" style="font-size:13px;"></i>
    Las matrices A y B deben tener formas compatibles para cada operación.
  </div>
</div>
```

```python
A = np.array([[20, 40], [30, 50]])
B = np.array([[4, 5], [2, 3]])
print("A:\n", A)
print("B:\n", B)

print("\nA + B:\n", A + B)
print("A - B:\n", A - B)
print("A * 5:\n", A * 5)
print("A * B (element-wise):\n", A * B)
print("A / 5:\n", A / 5)
print("A.dot(B) (producto matricial):\n", A.dot(B))
```
