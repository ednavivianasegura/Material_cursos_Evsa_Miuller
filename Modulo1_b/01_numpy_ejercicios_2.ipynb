```python
# 1. Suma de las matrices de junio y julio
suma = junio + julio
print("Suma Junio+Julio:\n", suma)

# 2. Transpuesta
T = suma.T
print("\nTranspuesta:\n", T)

# 3. Aplanar a vector
v = T.flatten()
print("\nVector:", v)

# 4. Suma acumulada
cs = v.cumsum()
print("\nCumsum:", cs)

# 5. Normalizar dividiendo entre el máximo
normalizado = cs / cs.max()
print("\nNormalizado (÷ máximo):", normalizado)
```

```python
# Más fácil

def matrix(n, limit):

    zero = np.zeros((n, n))

    for i in range(1, zero.shape[0]//2 - limit):

            if (i % 2) == 1:

                zero[i:-i, i:-i] = np.ones((n-i*2, n-i*2))

            else:

                zero[i:-i, i:-i] = np.zeros((n-i*2, n-i*2))

    return zero
```
