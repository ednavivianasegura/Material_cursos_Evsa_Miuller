```python
labels  = np.array(iris.columns[:-1])   # ["sepal_length", "sepal_width", "petal_length", "petal_width"]
print(labels)
pattern = iris.iloc[149, :-1]           # valores de la última flor (fila 149) — todas las columnas excepto "species"
print(pattern)


angles = np.linspace(0, 2*np.pi, len(labels), endpoint = False)  #  Calcular los ángulos Uen radianes

#Divide el círculo completo (2π radianes = 360°) en 4 partes iguales, una por cada variable.
#Con endpoint=False el último punto no repite el primero — los 4 ángulos quedan así:
#0°    → sepal_length  (derecha)
#90°   → sepal_width   (arriba)
#180°  → petal_length  (izquierda)
#270°  → petal_width   (abajo)
print(angles)

# Esto es para añadir el último lado del polígono:
# para Cerrar el polígono
pattern = np.concatenate((pattern, [pattern[0]]))
angles = np.concatenate((angles, [angles[0]]))

#Crear el gráfico polar
fig = plt.figure()
ax = fig.add_subplot(polar = True)

ax.plot(angles, pattern, "o-", linewidth = 4)           # Lineas en azul
ax.fill(angles, pattern, alpha = 1, color = "yellow")   # Relleno amarillo
ax.set_thetagrids(angles[:-1] * 180/np.pi, labels)      # Malla

plt.show()
```
