```python
# @title {"run":"auto","vertical-output":true,"single-column":true,"display-mode":"form"}
%%html

<link rel="stylesheet" href="https://cdn.jsdelivr.net/npm/@tabler/icons-webfont@latest/tabler-icons.min.css">
<style>
  @import url('https://fonts.googleapis.com/css2?family=Source+Code+Pro:wght@400;500&family=DM+Sans:wght@300;400;500;600&display=swap');
  .nb-cell{font-family:'DM Sans',sans-serif;padding:1.2rem 0;max-width:860px;}
  .nb-title{font-size:20px;font-weight:600;color:#1E3A8A;margin:0 0 .9rem 0;letter-spacing:-.3px;}
  .nb-title-sm{font-size:17px;font-weight:600;color:#1E3A8A;margin:0 0 .8rem 0;}
  .nb-badge{font-size:11px;font-weight:500;color:#0F766E;background:#CCFBF1;padding:3px 10px;border-radius:20px;border:1px solid #99F6E4;letter-spacing:.4px;text-transform:uppercase;}
  .nb-divider{height:2px;background:linear-gradient(90deg,#1E3A8A 0%,#475569 60%,transparent 100%);margin-bottom:1.25rem;border:none;}
  .nb-divider-sm{height:1px;background:linear-gradient(90deg,#475569 0%,transparent 80%);margin-bottom:1rem;border:none;}
  .nb-body{color:#1e293b;font-size:14px;line-height:1.75;margin-bottom:1rem;}
  .nb-body p{margin:0 0 .7rem;}
  .nb-body-box{font-size:13.5px;color:#1e293b;line-height:1.7;margin-bottom:1rem;padding:10px 14px;background:#F8FAFF;border-left:3px solid #1E3A8A;border-radius:0 6px 6px 0;}
  code.ic{font-family:'Source Code Pro',monospace;font-size:12.5px;background:#EFF6FF;color:#1E3A8A;padding:2px 7px;border-radius:4px;border:1px solid #BFDBFE;white-space:nowrap;}
  code.ic-g{font-family:'Source Code Pro',monospace;font-size:12.5px;background:#F0FDF4;color:#0F766E;padding:2px 7px;border-radius:4px;border:1px solid #99F6E4;white-space:nowrap;}
  code.ic-a{font-family:'Source Code Pro',monospace;font-size:12.5px;background:#FFF7ED;color:#92400E;padding:2px 7px;border-radius:4px;border:1px solid #FED7AA;white-space:nowrap;}
  .code-block{background:#0f172a;border-radius:8px;overflow:hidden;margin-top:.5rem;}
  .code-header{display:flex;align-items:center;justify-content:space-between;padding:7px 14px;background:#1e293b;border-bottom:1px solid #334155;}
  .code-lang{font-family:'Source Code Pro',monospace;font-size:11px;color:#64748b;letter-spacing:.5px;text-transform:uppercase;}
  .dots{display:flex;gap:5px;}.dot{width:9px;height:9px;border-radius:50%;}
  .dot-r{background:#ef4444;}.dot-y{background:#f59e0b;}.dot-g{background:#22c55e;}
  pre.cc{padding:14px 18px;font-family:'Source Code Pro',monospace;font-size:13px;line-height:1.7;color:#e2e8f0;margin:0;white-space:pre;}
  .kw{color:#7dd3fc;}.mod{color:#a5f3fc;}.al{color:#86efac;}.cm{color:#94A3B8;font-style:italic;}
  .fn{color:#c4b5fd;}.st{color:#86efac;}.num{color:#fcd34d;}.op{color:#f9a8d4;}
  .nb-note{display:flex;gap:10px;align-items:flex-start;background:#F0FDF4;border-left:3px solid #0F766E;border-radius:0 6px 6px 0;padding:10px 14px;margin-top:.8rem;}
  .nb-note-text{font-size:13px;color:#0f766e;line-height:1.6;}
  .warn-box{display:flex;align-items:flex-start;gap:10px;background:#FFF7ED;border-left:3px solid #F59E0B;border-radius:0 6px 6px 0;padding:10px 14px;margin-top:.8rem;}
  .warn-text{font-size:13px;color:#92400E;line-height:1.6;}
  .table-wrap{border-radius:8px;overflow:hidden;border:1px solid rgba(0,0,0,.08);box-shadow:0 2px 8px rgba(0,0,0,.05);}
  .ref-table{width:100%;border-collapse:collapse;font-size:13px;}
  .ref-table thead tr{background:#1E3A8A;color:#fff;}
  .ref-table thead th{padding:10px 16px;text-align:left;font-weight:500;font-size:11px;letter-spacing:.8px;text-transform:uppercase;}
  .ref-table thead th:first-child{border-radius:8px 0 0 0;width:35%;}
  .ref-table thead th:last-child{border-radius:0 8px 0 0;}
  .ref-table tbody tr:nth-child(odd){background:#F8FAFF;}
  .ref-table tbody tr:nth-child(even){background:#fff;}
  .ref-table tbody tr:hover{background:#EFF6FF;}
  .ref-table tbody td{padding:9px 16px;border-bottom:.5px solid rgba(0,0,0,.06);color:#1e293b;vertical-align:middle;}
  .ref-table tbody tr:last-child td{border-bottom:none;}
  .ref-table tbody tr:last-child td:first-child{border-radius:0 0 0 8px;}
  .ref-table tbody tr:last-child td:last-child{border-radius:0 0 8px 0;}
  .desc{font-size:13px;color:#475569;line-height:1.5;}
  .desc strong{color:#1E3A8A;font-weight:600;}
  .tip{position:relative;display:inline-block;}
  .tip .tt{visibility:hidden;opacity:0;background:#1e293b;color:#e2e8f0;font-family:'DM Sans',sans-serif;font-size:12px;line-height:1.5;border-radius:6px;padding:7px 11px;position:absolute;z-index:999;left:0;top:calc(100% + 6px);white-space:normal;max-width:300px;box-shadow:0 4px 12px rgba(0,0,0,.2);transition:opacity .18s ease;pointer-events:none;}
  .tip .tt::before{content:'';position:absolute;bottom:100%;left:14px;border:5px solid transparent;border-bottom-color:#1e293b;}
  .tip:hover .tt{visibility:visible;opacity:1;}
  .hint{font-size:11px;color:#94A3B8;margin-top:.6rem;display:flex;align-items:center;gap:5px;}
  .grid2{display:grid;grid-template-columns:1fr 1fr;gap:12px;margin-bottom:12px;}
  .card{background:#fff;border:1px solid rgba(0,0,0,.08);border-radius:10px;padding:1rem;}
  .card-title{font-size:11px;font-weight:600;letter-spacing:.8px;text-transform:uppercase;color:#475569;margin-bottom:.6rem;padding-bottom:.5rem;border-bottom:.5px solid rgba(0,0,0,.07);}
  .card-body{font-size:13px;color:#1e293b;line-height:1.6;}
  .rb{display:inline-block;font-size:10px;font-weight:600;padding:2px 7px;border-radius:20px;vertical-align:middle;white-space:nowrap;margin-left:5px;}
  .rb-blue{background:#EFF6FF;color:#1E3A8A;border:1px solid #BFDBFE;}
  .rb-teal{background:#F0FDF4;color:#0F766E;border:1px solid #99F6E4;}
  .rb-amber{background:#FFF7ED;color:#92400E;border:1px solid #FED7AA;}
</style>

<div class="nb-cell">
  <h3 class="nb-title">Filtros</h3>
  <hr class="nb-divider">
  <div class="nb-body-box">
    Aplicar un operador de comparación sobre una columna devuelve una <strong>Serie booleana</strong> (<code class="ic">True</code>/<code class="ic">False</code>). Usar esa Serie como índice del DataFrame filtra las filas que cumplen la condición.
    Para combinar varios filtros usar <code class="ic">&</code> (y) y <code class="ic">|</code> (o), agrupando cada condición entre paréntesis <code class="ic">()</code>.
  </div>

</div>
```

```python
# Para aplicar un filtro usamos los operadores de comparación

df["Largo Sepalo"] > 7
```

```python
# Esto retorna una Serie con Verdaderos y Falsos
# Si quisieramos aplicar ese filtro al DataFrame hariamos un "indexing" con el operador

df[df["Largo Sepalo"] > 7]

# Esto nos retorna el DataFrame solo con las filas que cumplen la condición
```

```python
# selecciona filas del DataFrame df donde el valor en la columna 'Largo Sepalo' es mayor que 7
# Ó el valor en la columna 'Ancho Petalo' es menor que 2.
# El operador | significa 'o' lógico.
df[(df["Largo Sepalo"] > 7) | (df["Ancho Petalo"] < 2)]
```

```python
# para aplicar más de un filtro
# El operador & significa 'y' lógico.
df[(df["Largo Sepalo"] > 7) & (df["Ancho Petalo"] == 2)]
```

```python
# Selecciona filas del DataFrame donde el valor en la columna `'Clase'` es igual a `'Iris-virginica'`
# Ó donde el valor en la columna `'Clase'` es igual a `'Iris-versicolor'`.
df[(df["Clase"] == "Iris-virginica") | (df["Clase"] == "Iris-versicolor")]
```

```python
df = pd.DataFrame({"Nombre" : ["Gael", "Paola", "Maria", "Josue", "Martha", "Romero", "Sandra"],
                   "Edad" : [27, 28, 29, 28, 30, 30, 28]})

df["Residencia"] = ["Madrid", "Madrid", "Bilbao", "Madrid", "Madrid", "Zaragoza", "?"]

df
```

```python
df.info()
```

```python
# Modificar el tipo de dato de la columna "edad" a "int8"

df["Edad"] = df["Edad"].astype("int8")

df.dtypes
```

```python
# Cambiar el tipo de dato de la columna "Nombre" a "category"

df["Nombre"] = df["Nombre"].astype("category")

df.dtypes
```

```python
# Ahora es tipo "category"

df["Nombre"]
```

```python
# Esta columna sigue siendo "object"

df["Residencia"]
```

```python
# Cambiar 2 o más a la vez

df[["Nombre", "Residencia"]] = df[["Nombre", "Residencia"]].astype("category")

df.info()
```

```python
df[df["Edad"].isin([30, 28])]
```

```python
# Si quisieramos filtrar por un rango, podemos usar: .between()
# Los dos extremos son incluidos
# Solo se puede usar en columnas númericas

df[(df["Edad"].between(28, 30))]
```

```python
# Con ~ podemos obtener el resultado opuesto:

df[~(df["Edad"].between(28, 29))]
```

```python
# Si tenemos más de 1 condición debemos agruparlas con parentesis ()

df[(df["Edad"].isin([30, 28])) & (df["Residencia"] != "Madrid")]
```

```python
df1 = pd.DataFrame({"Columna_1" : ["A", "B"],
                    "Columna_2" : ["C", "D"]})

df2 = pd.DataFrame({"Columna_3" : ["E", "F", "G"],
                    "Columna_4" : ["G", "H", "I"]})
```

```python
df1
```

```python
df2
```

```python
# Para concatenar horizontalmente hay que usar axis = 1
pd.concat([df1, df2], axis = 1)
```

```python
# Usando un solo pd.concat() podemos concatenar varios DataFrames a la vez, incluso repetirlos

pd.concat([df1, df2, df2, df2], axis = 1)
```

```python
#Vertical
df1 = pd.DataFrame({"Columna_1" : ["A", "B", "1"],
                    "Columna_2" : ["C", "D", "2"]})

df2 = pd.DataFrame({"Columna_1" : ["E", "F"],
                    "Columna_2" : ["G", "H"]})

# Ambos tienen el mismo nombre en las columnas
```

```python
df1
```

```python
df2
```

```python
# Como tienen el mismo nombre de columnas, pandas los agrupa automaticamente

pd.concat([df1, df2], axis = 0)
```

```python
df3 = pd.DataFrame({"Columna_1" : ["E", "F"],
                    "Columna_3" : ["G", "H"]})

df3
```

```python
# En este ejemplo los DataFrames tienen columnas diferentes

pd.concat([df1, df3], axis = 0)

# Por eso, pandas agrupa la columna en comun, y las que sean diferentes las llena con NaN's
```

```python
# Si nos fijamos, pandas incluso concatena los indices, para evitar este comportamiento podemos agregar otro parametro:

pd.concat([df1, df3], axis = 0, ignore_index = True)
```

```python
df1 = pd.DataFrame({"Nombre" : ["Gael", "Paola", "Maria", "Josue", "Martha", "Romero", "Sandra"],
                    "Edad" : [27, 28, 29, 28, 30, 30, 28]})
df1
```

```python
df2 = pd.DataFrame({"Nombre" : ["Gael", "Paola", "Maria", "Josue", "Martha", "Romero", "Sandro"],
                    "Residencia" : ["Madrid", "Madrid", "Bilbao", "Madrid", "Madrid", "Zaragoza", "Sevilla"]})
df2
```

```python
# pd.merge() Une dos DataFrames que tengan una columna en común:
# Por defecto "how" es "inner"

pd.merge(left = df1, right = df2, left_on = "Nombre", right_on = "Nombre", how = "inner")
```

```python
pd.merge(left = df1, right = df2, left_on = "Nombre", right_on = "Nombre", how = "outer")
```

```python
# pd.merge() usando otros parámetros:
# Nota: esto solo funciona si ambos DataFrames tienen la columna con el mismo nombre

pd.merge(left = df1, right = df2, on = "Nombre", how = "outer")
```
