```python
# @title {"run":"auto","vertical-output":true,"single-column":true,"display-mode":"form"}
%%html

<link rel="stylesheet" href="https://cdn.jsdelivr.net/npm/@tabler/icons-webfont@latest/tabler-icons.min.css">
<style>
  @import url('https://fonts.googleapis.com/css2?family=Source+Code+Pro:wght@400;500&family=DM+Sans:wght@300;400;500;600&display=swap');
  .nb-cell{font-family:'DM Sans',sans-serif;padding:1.2rem 0;max-width:860px;}
  .nb-title{font-size:20px;font-weight:600;color:#1E3A8A;margin:0 0 .9rem 0;letter-spacing:-.3px;}
  .nb-divider{height:2px;background:linear-gradient(90deg,#1E3A8A 0%,#475569 60%,transparent 100%);margin-bottom:1.25rem;border:none;}
  .nb-body-box{font-size:13.5px;color:#1e293b;line-height:1.7;margin-bottom:1rem;padding:10px 14px;background:#F8FAFF;border-left:3px solid #1E3A8A;border-radius:0 6px 6px 0;}
  code.ic{font-family:'Source Code Pro',monospace;font-size:12.5px;background:#EFF6FF;color:#1E3A8A;padding:2px 7px;border-radius:4px;border:1px solid #BFDBFE;white-space:nowrap;}
  .nb-note{display:flex;gap:10px;align-items:flex-start;background:#F0FDF4;border-left:3px solid #0F766E;border-radius:0 6px 6px 0;padding:10px 14px;margin-top:.8rem;}
  .nb-note-text{font-size:13px;color:#0f766e;line-height:1.6;}
  .grid2{display:grid;grid-template-columns:1fr 1fr;gap:12px;margin-bottom:1rem;}
  .card{background:#fff;border:1px solid rgba(0,0,0,.08);border-radius:10px;padding:1rem;}
  .card-title{font-size:11px;font-weight:600;letter-spacing:.8px;text-transform:uppercase;color:#475569;margin-bottom:.6rem;padding-bottom:.5rem;border-bottom:.5px solid rgba(0,0,0,.07);}
  .card-body{font-size:13px;color:#1e293b;line-height:1.6;}
  .tip{position:relative;display:inline-block;}
  .tip .tt{visibility:hidden;opacity:0;background:#1e293b;color:#e2e8f0;font-family:'DM Sans',sans-serif;font-size:12px;line-height:1.6;border-radius:8px;padding:10px 14px;position:absolute;z-index:999;left:0;top:calc(100% + 8px);width:290px;box-shadow:0 4px 16px rgba(0,0,0,.25);transition:opacity .2s ease;pointer-events:none;}
  .tip .tt::before{content:'';position:absolute;bottom:100%;left:14px;border:6px solid transparent;border-bottom-color:#1e293b;}
  .tip .tt strong{color:#7dd3fc;}
  .tip .tt code{font-family:monospace;background:#334155;padding:1px 5px;border-radius:3px;font-size:11.5px;}
  .tip-up .tt{top:auto;bottom:calc(100% + 8px);}
  .tip-up .tt::before{bottom:auto;top:100%;border-bottom-color:transparent;border-top-color:#1e293b;}
  .tip:hover .tt{visibility:visible;opacity:1;}
  .tip-word{color:#1E3A8A;font-weight:600;border-bottom:1.5px dashed #1E3A8A;cursor:help;}
</style>

<div class="nb-cell">
  <h3 class="nb-title">subplots() — Figura y ejes</h3>
  <hr class="nb-divider">

  <div class="nb-body-box">
    <span class="tip"><code class="ic">plt.subplots(filas, columnas)</code>
      <span class="tt">
        <strong>Sintaxis completa:</strong><br>
        <code>fig, ax = plt.subplots(1, 2, figsize=(10, 4), sharey=True)</code><br><br>
        · 1 fila, 2 columnas de subplots<br>
        · figsize en pulgadas (ancho, alto)<br>
        · sharey=True comparte el eje Y entre subplots
      </span>
    </span>
    crea una <strong>figura</strong> (<code class="ic">fig</code>) con una cuadrícula de
    <strong>ejes</strong> (<code class="ic">ax</code>). Es la forma moderna y recomendada
    de trabajar con matplotlib. Con un solo subplot <code class="ic">ax</code> es un único
    objeto; con múltiples, <code class="ic">ax</code> es un array accesible con
    <span class="tip"><code class="ic">ax[i]</code>
      <span class="tt">
        <strong>Acceso a subplots:</strong><br>
        · 1 fila → <code>ax[0]</code>, <code>ax[1]</code>…<br>
        · 2D (filas y columnas) → <code>ax[0, 1]</code><br><br>
        Ejemplo con 1×3:<br>
        <code>fig, ax = plt.subplots(1, 3)</code><br>
        <code>ax[0].plot(...)</code> → primer subplot<br>
        <code>ax[2].plot(...)</code> → tercer subplot
      </span>
    </span>
    (1D) o <code class="ic">ax[i, j]</code> (2D).
  </div>

  <div class="grid2">
    <div class="card">
      <div class="card-title"><i class="ti ti-file" style="font-size:13px;color:#1E3A8A;vertical-align:-1px;"></i> &nbsp;fig — Figura</div>
      <div class="card-body">El <strong>lienzo completo</strong> que contiene todos los ejes. Se usa para ajustes globales:
        <span class="tip tip-up"><code class="ic">fig.suptitle()</code>
          <span class="tt">
            <strong>fig.suptitle()</strong><br>
            Título general para toda la figura — aparece encima de todos los subplots.<br><br>
            <code>fig.suptitle("Mi título", fontsize=14)</code>
          </span>
        </span>,
        <span class="tip tip-up"><code class="ic">fig.tight_layout()</code>
          <span class="tt">
            <strong>fig.tight_layout()</strong><br>
            Ajusta automáticamente los márgenes entre subplots para evitar que los títulos y etiquetas se solapen.<br><br>
            Siempre llamarlo <strong>antes</strong> de <code>plt.show()</code> y <code>fig.savefig()</code>.
          </span>
        </span> y
        <span class="tip tip-up"><code class="ic">fig.savefig()</code>
          <span class="tt">
            <strong>fig.savefig()</strong><br>
            Guarda la figura en disco. Debe llamarse <strong>antes</strong> de <code>plt.show()</code>, o la figura guardada aparecerá en blanco.<br><br>
            <code>fig.savefig("figura.png", dpi=300)</code><br>
            <code>fig.savefig("figura.pdf")</code>
          </span>
        </span>.
      </div>
    </div>
    <div class="card">
      <div class="card-title"><i class="ti ti-chart-line" style="font-size:13px;color:#1E3A8A;vertical-align:-1px;"></i> &nbsp;ax — Axes</div>
      <div class="card-body">Cada <strong>recuadro de dibujo</strong> individual. Se usa para
        <span class="tip tip-up"><code class="ic">ax.plot()</code>
          <span class="tt">
            <strong>API de Axes (recomendada)</strong><br>
            En lugar de <code>plt.title()</code> → <code>ax.set_title()</code><br>
            En lugar de <code>plt.xlabel()</code> → <code>ax.set_xlabel()</code><br>
            En lugar de <code>plt.ylabel()</code> → <code>ax.set_ylabel()</code><br>
            En lugar de <code>plt.legend()</code> → <code>ax.legend()</code>
          </span>
        </span>,
        <code class="ic">ax.set_title()</code>, <code class="ic">ax.set_xlabel()</code>…
        Admite
        <span class="tip tip-up"><code class="ic">sharey=True</code>
          <span class="tt">
            <strong>sharey / sharex</strong><br>
            Comparte el eje Y (o X) entre todos los subplots.<br>
            Facilita la comparación visual entre gráficos al mantener la misma escala.<br><br>
            <code>fig, ax = plt.subplots(1, 3, sharey=True)</code>
          </span>
        </span>
        para compartir el eje Y entre subplots.
      </div>
    </div>
  </div>

  <div class="nb-note">
    <i class="ti ti-info-circle" style="color:#0F766E;font-size:15px;flex-shrink:0;margin-top:1px;"></i>
    <span class="nb-note-text">
      <code class="ic">fig.tight_layout()</code> ajusta automáticamente los márgenes para evitar solapamientos.
      <code class="ic">fig.savefig()</code> debe llamarse <strong>antes</strong> de <code class="ic">plt.show()</code>,
      o la figura guardada aparecerá en blanco.
    </span>
  </div>
</div>
```

```python
# Creamos la figura "fig" donde se le van agregar 3 subplots utilizando la variable "ax"

fig, ax = plt.subplots(nrows = 3, ncols = 2, figsize = (12, 12))

# Primera Columna
X = np.linspace(0, 10, 100)

ax[0, 0].plot(X, np.sin(X))

ax[1, 0].plot(X, np.cos(X))

ax[2, 0].plot(X, np.tan(X))

# Segunda Columna
Y = np.linspace(-1, 1, 100)

ax[0, 1].plot(X, np.arcsin(Y))

ax[1, 1].plot(X, np.arccos(Y))

ax[2, 1].plot(X, np.arctan(Y))

plt.show()
```

```python
# Utilizamos un bucle for para llenar cada uno de los subplots de la figura

x = np.linspace(-10, 10, 20)
y = x

fig, ax = plt.subplots(nrows = 2, ncols = 3, figsize = (14, 10)) #pulgadas
k = 1

for rows in ax:
    for ax1 in rows:
        ax1.plot(x, y**k, "blue")
        k += 1

plt.show()
```

```python
# Función de Schwefel — superficie 3D con mínimo global en (420.97, 420.97)

fig = plt.figure(figsize=(12, 8))

# add_subplot con projection="3d" activa el sistema de coordenadas tridimensional
ax = fig.add_subplot(projection="3d")

# np.arange genera valores de -100 a 100 con paso 0.1
# → 2000 puntos por eje
X = np.arange(-100, 100, 0.1)
Y = np.arange(-100, 100, 0.1)

# meshgrid transforma los dos vectores 1D en dos matrices 2D
# que representan todas las combinaciones posibles de (x, y)
# X e Y pasan de forma (2000,) a forma (2000, 2000)
X, Y = np.meshgrid(X, Y)

# Ecuación de Schwefel:
# f(x, y) = 418.9829·2 - x·sen(√|x|) - y·sen(√|y|)
# El mínimo global es 0 cuando x = y = 420.9687
# El factor 418.9829·2 desplaza la función para que el mínimo sea 0
Z = 418.9829 * 2 - (X * np.sin(np.sqrt(abs(X))) + Y * np.sin(np.sqrt(abs(Y))))

# plot_surface dibuja la superficie 3D a partir de las matrices X, Y, Z
# Cada celda de la malla es un pequeño rectángulo coloreado
ax.plot_surface(X, Y, Z)

plt.show()
```
