```python
# ════════════════════════════════════════════════════════
# plt.plot() — línea con parámetros de estilo
# ════════════════════════════════════════════════════════
x = np.linspace(0, 2 * np.pi, 100)

# ┌──────────────────────────────────────────────────────┐
# │  🟣 CAMBIA ESTOS PARÁMETROS para ver el efecto:     │
# │   color  →  'red', 'green', '#FF6600'               │
# │   ls     →  '-', '-.', ':'  (estilo de línea)       │
# │   lw     →  0.5, 4          (grosor)                │
# │   marker →  'o', 's', '^', '*', '+'                 │
# │   ms     →  2, 8            (tamaño del marcador)   │
# └──────────────────────────────────────────────────────┘
plt.plot(x, np.sin(x), color='royalblue', ls='--', lw=2, marker='o', ms=4, label='sen(x)')
plt.plot(x, np.cos(x), color='tomato',    ls='-',  lw=1.5,             label='cos(x)')

plt.title('plt.plot() — dos curvas en la misma figura')
plt.xlabel('x (radianes)')  ;  plt.ylabel('Amplitud')
plt.legend()
# ┌──────────────────────────────────────────────────────┐
# │  🟣 CAMBIA True por False para quitar la cuadrícula  │
# └──────────────────────────────────────────────────────┘
plt.grid(True, linestyle=':', alpha=0.4)
plt.show()
```

```python
# ════════════════════════════════════════════════════════
# sns.set_theme() — cambia el aspecto de TODOS los gráficos
# ════════════════════════════════════════════════════════

# ┌──────────────────────────────────────────────────────┐
# │  🟣 CAMBIA 'whitegrid' a:                           │
# │   'darkgrid'  → fondo oscuro con cuadrícula         │
# │   'ticks'     → solo marcas, sin cuadrícula         │
# │   'white'     → fondo blanco limpio                 │
# └──────────────────────────────────────────────────────┘
sns.set_theme(style='whitegrid')
print('Tema aplicado: whitegrid — afecta a todos los gráficos siguientes')
```

```python
# ════════════════════════════════════════════════════════
# sns.countplot() — frecuencias de una variable categórica
# ════════════════════════════════════════════════════════

# ┌──────────────────────────────────────────────────────┐
# │  🟣 CAMBIA hue='survived' a hue='sex'               │
# │  CAMBIA palette='Set2' a 'Set1', 'pastel', 'deep'   │
# │  CAMBIA x='pclass' a x='sex' para otro eje         │
# └──────────────────────────────────────────────────────┘
sns.countplot(data=titanic, x='pclass', hue='survived', palette='Set2')
plt.title('countplot — Pasajeros por clase y supervivencia')
plt.xlabel('Clase')  ;  plt.legend(title='Sobrevivió')  ;  plt.show()
```

```python
# ════════════════════════════════════════════════════════
# sns.histplot() — histograma con KDE
# Comparación: Matplotlib puro vs Seaborn
# ════════════════════════════════════════════════════════
titanic_clean = titanic.dropna(subset=['age'])
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 4))

# Matplotlib puro — sin KDE
ax1.hist(titanic_clean['age'], bins=25, color='steelblue', alpha=0.7, edgecolor='white')
ax1.set_title('Matplotlib — hist()')  ;  ax1.set_xlabel('Edad')

# Seaborn — KDE integrada con una sola línea
# ┌──────────────────────────────────────────────────────┐
# │  🟣 CAMBIA kde=True a False para quitar la curva    │
# │  CAMBIA element='bars' a 'step' o 'poly'            │
# │  AÑADE hue='survived' para colorear por grupo       │
# └──────────────────────────────────────────────────────┘
sns.histplot(data=titanic_clean, x='age', bins=25,
             kde=True, element='bars', color='salmon', ax=ax2)
ax2.set_title('Seaborn — histplot(kde=True)')  ;  ax2.set_xlabel('Edad')

plt.suptitle('Histograma: Matplotlib vs Seaborn', fontsize=13, fontweight='bold')
plt.tight_layout()  ;  plt.show()
```

```python
# ════════════════════════════════════════════════════════
# sns.scatterplot() — hue + style + size en una línea
# ════════════════════════════════════════════════════════

# ┌──────────────────────────────────────────────────────┐
# │  🟣 PRUEBA quitar un parámetro cada vez:            │
# │   size=None    → todos los puntos del mismo tamaño  │
# │   style=None   → todos los puntos con el mismo form │
# │   hue=None     → sin color por categoría            │
# │  CAMBIA palette='deep' a 'Set1', 'tab10'           │
# └──────────────────────────────────────────────────────┘
plt.figure(figsize=(8, 5))
sns.scatterplot(data=iris_sns, x='sepal_length', y='sepal_width',
                hue='species', style='species', size='petal_width',
                palette='deep', sizes=(20, 200))
plt.title('scatterplot — hue + style + size')  ;  plt.tight_layout()  ;  plt.show()
```

```python
# ════════════════════════════════════════════════════════
# sns.jointplot() — scatter central + distribuciones marginales
# ════════════════════════════════════════════════════════

# ┌──────────────────────────────────────────────────────┐
# │  🟣 CAMBIA kind='scatter' a:                        │
# │   'kde'  → curvas de densidad suavizadas            │
# │   'hex'  → hexágonos (útil con muchos datos)        │
# │  CAMBIA hue='species' a None para sin colores       │
# └──────────────────────────────────────────────────────┘
g = sns.jointplot(data=iris_sns, x='sepal_length', y='petal_length',
                  hue='species', kind='scatter', palette='Set1')
g.figure.suptitle('jointplot — distribuciones marginales', y=1.02)
plt.show()
```

```python
# ════════════════════════════════════════════════════════
# sns.pairplot() — matriz de todos los pares de variables
# ════════════════════════════════════════════════════════

# ┌──────────────────────────────────────────────────────┐
# │  🟣 CAMBIA diag_kind='hist' a 'kde'                 │
# │  CAMBIA corner=False a True → solo mitad inferior   │
# └──────────────────────────────────────────────────────┘
g = sns.pairplot(iris_sns, hue='species', diag_kind='hist',
                 palette='Set2', corner=False)
g.figure.suptitle('Pairplot — Iris por especie', y=1.02, fontsize=13)
plt.show()
```

```python
# ════════════════════════════════════════════════════════
# kdeplot, boxplot y violinplot — comparación
# ════════════════════════════════════════════════════════
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

# [0] KDE por especie
for especie, color in zip(iris_sns['species'].unique(), ['steelblue','salmon','seagreen']):
    # ┌────────────────────────────────────────────────┐
    # │  🟣 CAMBIA fill=True a False → solo curva     │
    # │  bw_adjust=1.0 → más bajo=más picuda           │
    # └────────────────────────────────────────────────┘
    sns.kdeplot(data=iris_sns[iris_sns['species']==especie], x='petal_length',
                fill=True, alpha=0.4, color=color, label=especie, bw_adjust=1.0, ax=axes[0])
axes[0].set_title('kdeplot')  ;  axes[0].legend()

# [1] Boxplot Seaborn
# ┌──────────────────────────────────────────────────────┐
# │  🟣 CAMBIA palette='Set2' a 'pastel', 'deep'        │
# └──────────────────────────────────────────────────────┘
sns.boxplot(data=iris_sns, x='species', y='sepal_length',
            hue='species', palette='Set2', ax=axes[1], legend=False)
axes[1].set_title('boxplot()')

# [2] Violinplot
# ┌──────────────────────────────────────────────────────┐
# │  🟣 CAMBIA inner='box' a 'quartile', 'point', None  │
# └──────────────────────────────────────────────────────┘
sns.violinplot(data=iris_sns, x='species', y='sepal_length',
               hue='species', palette='Set1', inner='box', ax=axes[2], legend=False)
axes[2].set_title('violinplot()')

plt.suptitle('kdeplot · boxplot · violinplot', fontsize=13, fontweight='bold')
plt.tight_layout()  ;  plt.show()
```

```python
# ════════════════════════════════════════════════════════
# sns.heatmap() — correlación con df.corr()
# ════════════════════════════════════════════════════════
corr = iris_sns.select_dtypes('number').corr()

# ┌──────────────────────────────────────────────────────┐
# │  🟣 CAMBIA cmap='coolwarm' a 'viridis', 'RdBu'     │
# │  CAMBIA annot=True a False para quitar los números  │
# │  ELIMINA vmin=-1, vmax=1 para rango libre           │
# └──────────────────────────────────────────────────────┘
plt.figure(figsize=(6, 5))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm',
            linewidths=0.5, vmin=-1, vmax=1)
plt.title('Heatmap — correlación Iris')  ;  plt.show()
```

```python
# ════════════════════════════════════════════════════════
# sns.heatmap() sobre pivot_table — media por 2 categorías
# Patrón: df.pivot_table(values=, index=, columns=, aggfunc=)
# ════════════════════════════════════════════════════════

# ┌──────────────────────────────────────────────────────┐
# │  🟣 CAMBIA values='fare' a 'age'                    │
# │  CAMBIA aggfunc='mean' a 'count' o 'sum'            │
# └──────────────────────────────────────────────────────┘
pivot = titanic.pivot_table(values='fare',
                             index='pclass',
                             columns='sex',
                             aggfunc='mean')
print('Tabla pivot — precio medio por clase y sexo:')
print(pivot.round(2))

# ┌──────────────────────────────────────────────────────┐
# │  🟣 CAMBIA cmap='YlOrRd' a 'Blues', 'coolwarm'     │
# │  fmt='.1f' = 1 decimal. Prueba '.0f' (sin decimales)│
# └──────────────────────────────────────────────────────┘
plt.figure(figsize=(5, 4))
sns.heatmap(pivot, annot=True, fmt='.1f', cmap='YlOrRd', linewidths=0.5)
plt.title('Heatmap — Precio medio del billete por clase y sexo')
plt.show()
```

```python
# ════════════════════════════════════════════════════════
# sns.FacetGrid() — un panel por categoría
# Patrón: g = sns.FacetGrid(df, col='cat') ; g.map(fn, 'var')
# ════════════════════════════════════════════════════════
titanic_clean = titanic.dropna(subset=['age'])

# ┌──────────────────────────────────────────────────────┐
# │  🟣 CAMBIA col='survived' a col='pclass'            │
# │  → obtienes un panel por clase (1, 2 y 3)           │
# │  CAMBIA sns.histplot a sns.kdeplot                  │
# └──────────────────────────────────────────────────────┘
g = sns.FacetGrid(titanic_clean, col='survived', height=4)
g.map(sns.histplot, 'age', bins=20, color='steelblue')
g.set_axis_labels('Edad', 'Frecuencia')
g.set_titles(col_template='Survived = {col_name}')
plt.suptitle('FacetGrid — un panel por valor de survived', y=1.05)
plt.show()
```

```python
# ════════════════════════════════════════════════════════
# FacetGrid con col= Y row= → cuadrícula 2D de paneles
# ════════════════════════════════════════════════════════

# ┌──────────────────────────────────────────────────────┐
# │  col='sex'      → columnas: female / male           │
# │  row='survived' → filas: 0 (falleció) / 1 (vivió)  │
# │  → 4 paneles en total                               │
# │  🟣 CAMBIA row='survived' a row='pclass'            │
# │  para obtener una cuadrícula 2×3                    │
# └──────────────────────────────────────────────────────┘
g2 = sns.FacetGrid(titanic_clean, col='sex', row='survived', height=3)
g2.map(sns.kdeplot, 'age', fill=True, color='salmon')
g2.set_axis_labels('Edad', 'Densidad')
plt.suptitle('FacetGrid con col= y row= → cuadrícula 2D', y=1.02)
plt.show()
```

```python
# ════════════════════════════════════════════════════════
# Radar chart — matplotlib con projection='polar'
# (Seaborn no lo tiene nativo)
# ════════════════════════════════════════════════════════
labels  = np.array(iris_sns.columns[:-1])   # 4 variables numéricas
angles  = np.linspace(0, 2*np.pi, len(labels), endpoint=False)

fig = plt.figure(figsize=(7, 7))
ax  = fig.add_subplot(polar=True)   # projection='polar' es la clave

# ┌──────────────────────────────────────────────────────┐
# │  🟣 CAMBIA los índices 0, 50, 100 a otras filas     │
# │  por ejemplo 10, 75, 130                            │
# │  CAMBIA alpha=0.15 a 0.4 → relleno más opaco        │
# └──────────────────────────────────────────────────────┘
for idx, color, label in zip([0, 50, 100],
                              ['royalblue', 'tomato', 'seagreen'],
                              ['Setosa (fila 0)', 'Versicolor (fila 50)', 'Virginica (fila 100)']):
    pattern = iris_sns.iloc[idx, :-1].values
    p = np.concatenate((pattern, [pattern[0]]))   # cerrar el polígono (último == primero)
    a = np.concatenate((angles,  [angles[0]]))
    ax.plot(a, p, 'o-', lw=2, color=color, label=label)
    ax.fill(a, p, alpha=0.15, color=color)

ax.set_thetagrids(angles * 180/np.pi, labels)   # etiquetas en los ángulos
ax.set_title('Radar chart — comparación de 3 flores', pad=20)
ax.legend(loc='upper right', bbox_to_anchor=(1.45, 1.1))
plt.show()
```
