```python
# @title
%%html

<link rel="stylesheet" href="https://cdn.jsdelivr.net/npm/@tabler/icons-webfont@latest/tabler-icons.min.css">
<style>
  @import url('https://fonts.googleapis.com/css2?family=Source+Code+Pro:wght@400;500&family=DM+Sans:wght@300;400;500;600&display=swap');
  .nb-cell{font-family:'DM Sans',sans-serif;padding:1.2rem 0;max-width:860px;}
  .nb-title{font-size:20px;font-weight:600;color:#1E3A8A;margin:0 0 .9rem 0;letter-spacing:-.3px;}
  .nb-divider{height:2px;background:linear-gradient(90deg,#1E3A8A 0%,#475569 60%,transparent 100%);margin-bottom:1.25rem;border:none;}
  .nb-body-box{font-size:13.5px;color:#1e293b;line-height:1.7;margin-bottom:1rem;padding:10px 14px;background:#F8FAFF;border-left:3px solid #1E3A8A;border-radius:0 6px 6px 0;}
  code.ic{font-family:'Source Code Pro',monospace;font-size:12.5px;background:#EFF6FF;color:#1E3A8A;padding:2px 7px;border-radius:4px;border:1px solid #BFDBFE;white-space:nowrap;}
  code.ic-g{font-family:'Source Code Pro',monospace;font-size:12.5px;background:#F0FDF4;color:#0F766E;padding:2px 7px;border-radius:4px;border:1px solid #99F6E4;white-space:nowrap;}
  .code-block{background:#0f172a;border-radius:8px;overflow:hidden;margin-top:.5rem;}
  .code-header{display:flex;align-items:center;justify-content:space-between;padding:7px 14px;background:#1e293b;border-bottom:1px solid #334155;}
  .code-lang{font-family:'Source Code Pro',monospace;font-size:11px;color:#64748b;letter-spacing:.5px;text-transform:uppercase;}
  .dots{display:flex;gap:5px;}.dot{width:9px;height:9px;border-radius:50%;}
  .dot-r{background:#ef4444;}.dot-y{background:#f59e0b;}.dot-g{background:#22c55e;}
  pre.cc{padding:14px 18px;font-family:'Source Code Pro',monospace;font-size:13px;line-height:1.7;color:#e2e8f0;margin:0;white-space:pre;}
  .kw{color:#7dd3fc;}.mod{color:#a5f3fc;}.al{color:#86efac;}.cm{color:#94A3B8;font-style:italic;}
  .fn{color:#c4b5fd;}.st{color:#86efac;}.num{color:#fcd34d;}.op{color:#f9a8d4;}
  .nb-note{display:flex;gap:10px;align-items:flex-start;background:#F0FDF4;border-left:3px solid #0F766E;border-radius:0 6px 6px 0;padding:10px 14px;margin-top:.8rem;}
  .nb-note-text{font-size:13px;color:#0f766e;line-height:1.6;}
  .warn-box{display:flex;align-items:flex-start;gap:10px;background:#FFF7ED;border-left:3px solid #F59E0B;border-radius:0 6px 6px 0;padding:10px 14px;margin-top:.8rem;}
  .warn-text{font-size:13px;color:#92400E;line-height:1.6;}
  .table-wrap{border-radius:8px;overflow:hidden;border:1px solid rgba(0,0,0,.08);box-shadow:0 2px 8px rgba(0,0,0,.05);}
  .ref-table{width:100%;border-collapse:collapse;font-size:13px;}
  .ref-table thead tr{background:#1E3A8A;color:#fff;}
  .ref-table thead th{padding:10px 16px;text-align:left;font-weight:500;font-size:11px;letter-spacing:.8px;text-transform:uppercase;}
  .ref-table thead th:first-child{border-radius:8px 0 0 0;width:30%;}
  .ref-table thead th:last-child{border-radius:0 8px 0 0;}
  .ref-table tbody tr:nth-child(odd){background:#F8FAFF;}
  .ref-table tbody tr:nth-child(even){background:#fff;}
  .ref-table tbody tr:hover{background:#EFF6FF;}
  .ref-table tbody td{padding:9px 16px;border-bottom:.5px solid rgba(0,0,0,.06);color:#1e293b;vertical-align:middle;}
  .ref-table tbody tr:last-child td{border-bottom:none;}
  .ref-table tbody tr:last-child td:first-child{border-radius:0 0 0 8px;}
  .ref-table tbody tr:last-child td:last-child{border-radius:0 0 8px 0;}
  .desc{font-size:13px;color:#475569;line-height:1.5;}
  .desc strong{color:#1E3A8A;font-weight:600;}
  .tip{position:relative;display:inline-block;}
  .tip .tt{visibility:hidden;opacity:0;background:#1e293b;color:#e2e8f0;font-family:'DM Sans',sans-serif;font-size:12px;line-height:1.5;border-radius:6px;padding:7px 11px;position:absolute;z-index:999;left:0;top:calc(100% + 6px);white-space:normal;max-width:300px;box-shadow:0 4px 12px rgba(0,0,0,.2);transition:opacity .18s ease;pointer-events:none;}
  .tip .tt::before{content:'';position:absolute;bottom:100%;left:14px;border:5px solid transparent;border-bottom-color:#1e293b;}
  .tip:hover .tt{visibility:visible;opacity:1;}
  .hint{font-size:11px;color:#94A3B8;margin-top:.6rem;display:flex;align-items:center;gap:5px;}
  .grid2{display:grid;grid-template-columns:1fr 1fr;gap:12px;margin-bottom:12px;}
  .card{background:#fff;border:1px solid rgba(0,0,0,.08);border-radius:10px;padding:1rem;}
  .card-title{font-size:11px;font-weight:600;letter-spacing:.8px;text-transform:uppercase;color:#475569;margin-bottom:.6rem;padding-bottom:.5rem;border-bottom:.5px solid rgba(0,0,0,.07);}
  .card-body{font-size:13px;color:#1e293b;line-height:1.6;}
</style>

<div class="nb-cell">
  <h3 class="nb-title">Series.map()</h3>
  <hr class="nb-divider">
 <div class="nb-body-box">
    <code class="ic">.map()</code> aplica una función a <strong>cada elemento</strong> de una
    Serie o DataFrame. Acepta una <strong>función lambda</strong>, una <strong>función normal</strong>
    o un <strong>diccionario</strong> como parámetro, y <strong>no es in-place</strong> —
    devuelve un nuevo objeto. El parámetro opcional <code class="ic">na_action='ignore'</code>
    permite omitir los <code class="ic">NaN</code> durante la transformación.
    </div>
  <div class="table-wrap">
    <table class="ref-table">
      <thead><tr>
        <th><i class="ti ti-function"></i> &nbsp;Uso</th>
        <th><i class="ti ti-align-left"></i> &nbsp;Descripción</th>
      </tr></thead>
      <tbody>
        <tr>
          <td><span class="tip"><code class="ic">.map(lambda x: f(x))</code><span class="tt">La función recibe cada elemento de la Serie como argumento x y devuelve el valor transformado.</span></span></td>
          <td class="desc">Aplica una <strong>función lambda</strong> elemento a elemento</td>
        </tr>
        <tr>
          <td><span class="tip"><code class="ic">.map(&#123;&quot;a&quot;: 1, &quot;b&quot;: 2&#125;)</code><span class="tt">Equivalente a .replace() con diccionario. Los valores que no están en el diccionario se convierten en NaN.</span></span></td>
          <td class="desc">Mapea valores usando un <strong>diccionario</strong></td>
        </tr>
      </tbody>
    </table>
  </div>

  <div class="nb-note" style="margin-top:.8rem;">
    <i class="ti ti-info-circle" style="color:#0F766E;font-size:15px;flex-shrink:0;margin-top:1px;"></i>
    <span class="nb-note-text">Cuando se usa con diccionario, los valores que <strong>no aparecen como clave</strong> en el diccionario se convierten en <code class="ic">NaN</code>. Si se quiere conservarlos, es preferible usar <code class="ic">.replace()</code>.</span>
  </div>
</div>
```

```python
df["Age"]
```

```python
# Raíz cuadrada de la edad
df["Age"].map(lambda x: np.sqrt(x))
```

```python
# Multiplicar por 2
df["Age"].map(lambda x: x * 2)
```

```python
# Guardar en nueva columna
df["AgeSQRT"] = df["Age"].map(lambda x: np.sqrt(x))
df.head()
```

```python
# La función lambda nos deja aplicar muchas funciones sin tener que definir una
# Extraer apellido
df["Surname"] = df["Name"].map(lambda x: x.split(",")[0])
df.head()
```

```python
# Condicional en lambda
df["AgeSQRT"].map(lambda x: "mayor" if x > 5 else "menor")
```

```python
# Con diccionario
df["Sex"].map({"male": "H", "female": "M"})
```

```python
# .groupby() se usa para agrupar filas que tienen los mismos valores.
# Obligatoriamente se usa junto con funciones agregadas para producir informes resumidos.

df.groupby("Continente")

# Me devuelve el objeto de la agrupación
```

```python
# .first() Me devuelve un elemento de cada agrupación

df.groupby("Continente").first()
```

```python
# .size() Devuelve el tamaño de cada grupo de la agrupación

df.groupby(["Continente"]).size()
```

```python
# Podemos agrupar por más de una columna

df.groupby(["Continente", "Tipo de region"]).first()
```

```python
df.groupby(["Tipo de region", "Continente"]).first()
```

```python
# .index nos retorna la forma del indice del DataFrame

df.groupby(["Tipo de region", "Continente"]).first().index
```

```python
# .last() Devuelve el último elemento de la agrupación
df.groupby(["Continente"]).last()
```

```python
#  .count() Cuenta el número de elementos que no son NaN's

df.groupby(["Continente"]).count()
```

```python
# Agrupamos por continente y luego sumamos la migración de cada continente en el año 1980

df.groupby("Continente")[1980].sum()
```

```python
# Equivalente con bucle for
dict_cont = {}
for cont, valor in df[["Continente", 1980]].values:
    if cont not in dict_cont.keys():
        dict_cont[cont] = valor
    else:
        dict_cont[cont] = dict_cont[cont] + valor
dict_cont
```

```python
df.groupby("Continente")[[2008, 2009, 2010]].sum()
```

```python
# Para encontrar los valores máximos de cada continente en el año 1980

df.groupby("Continente")[1980].max()
```

```python
# .aggregate() toma una lista de funciones agregadas y las aplica a todas las columnas

df.groupby("Continente")[[2010, 2011, 2012]].aggregate(["min", "max", "sum"])
```

```python
df.groupby("Continente")[list(range(1980, 2014))].aggregate(["min", "max", "sum"])
```

```python
# .agg() toma como parametro un diccionario
# Las llaves son los nombres de las columnas
# Los valores son listas de funciones agregadas
# Aquí podemos elegir a que columnas aplicar que funcion

df.groupby("Continente").agg({2010 : ["min", "max", "sum", "std"],
                              2011 : ["min", "max", "mean"],
                              2012 : ["sum", "median"]})
```

```python
df5 = df.groupby("Continente").agg({2010 : ["min","max","sum"],
                                    2011 : ["min","max"],
                                    2012 : ["sum"]})

df5.reset_index()
```

```python
# Para tener un resultados como el anterior, podemos pasarle un parametro extra a .groupby()

df.groupby("Continente", as_index = False).agg({2010 : ["min","max","sum"],
                                                2011 : ["min","max"],
                                                2012 : ["sum"]})
```

```python
# @title
%%html

<link rel="stylesheet" href="https://cdn.jsdelivr.net/npm/@tabler/icons-webfont@latest/tabler-icons.min.css">
<style>
  @import url('https://fonts.googleapis.com/css2?family=Source+Code+Pro:wght@400;500&family=DM+Sans:wght@300;400;500;600&display=swap');
  .nb-cell{font-family:'DM Sans',sans-serif;padding:1.2rem 0;max-width:860px;}
  .nb-title{font-size:20px;font-weight:600;color:#1E3A8A;margin:0 0 .9rem 0;letter-spacing:-.3px;}
  .nb-divider{height:2px;background:linear-gradient(90deg,#1E3A8A 0%,#475569 60%,transparent 100%);margin-bottom:1.25rem;border:none;}
  .nb-body-box{font-size:13.5px;color:#1e293b;line-height:1.7;margin-bottom:1rem;padding:10px 14px;background:#F8FAFF;border-left:3px solid #1E3A8A;border-radius:0 6px 6px 0;}
  code.ic{font-family:'Source Code Pro',monospace;font-size:12.5px;background:#EFF6FF;color:#1E3A8A;padding:2px 7px;border-radius:4px;border:1px solid #BFDBFE;white-space:nowrap;}
  .code-block{background:#0f172a;border-radius:8px;overflow:hidden;margin-top:.5rem;}
  .code-header{display:flex;align-items:center;justify-content:space-between;padding:7px 14px;background:#1e293b;border-bottom:1px solid #334155;}
  .code-lang{font-family:'Source Code Pro',monospace;font-size:11px;color:#64748b;letter-spacing:.5px;text-transform:uppercase;}
  .dots{display:flex;gap:5px;}.dot{width:9px;height:9px;border-radius:50%;}
  .dot-r{background:#ef4444;}.dot-y{background:#f59e0b;}.dot-g{background:#22c55e;}
  pre.cc{padding:14px 18px;font-family:'Source Code Pro',monospace;font-size:13px;line-height:1.7;color:#e2e8f0;margin:0;white-space:pre;}
  .kw{color:#7dd3fc;}.mod{color:#a5f3fc;}.al{color:#86efac;}.cm{color:#94A3B8;font-style:italic;}
  .fn{color:#c4b5fd;}.st{color:#86efac;}.num{color:#fcd34d;}.op{color:#f9a8d4;}
  .nb-note{display:flex;gap:10px;align-items:flex-start;background:#F0FDF4;border-left:3px solid #0F766E;border-radius:0 6px 6px 0;padding:10px 14px;margin-top:.8rem;}
  .nb-note-text{font-size:13px;color:#0f766e;line-height:1.6;}
  .warn-box{display:flex;align-items:flex-start;gap:10px;background:#FFF7ED;border-left:3px solid #F59E0B;border-radius:0 6px 6px 0;padding:10px 14px;margin-top:.8rem;}
  .warn-text{font-size:13px;color:#92400E;line-height:1.6;}
  .nb-divider-sm{height:1px;background:linear-gradient(90deg,#475569 0%,transparent 80%);margin:1rem 0;border:none;}
</style>

<div class="nb-cell">
  <h3 class="nb-title">Eliminar multi-índice en columnas</h3>
  <hr class="nb-divider">

  <div class="nb-body-box">
    Cuando se usa <code class="ic">.agg()</code> con varias funciones, pandas genera un
    <strong>multi-índice</strong> en las columnas (ej: <code class="ic">(2010, "min")</code>,
    <code class="ic">(2010, "max")</code>…). Hay dos formas de resolverlo:
    aplanar el resultado con una list comprehension, o evitarlo desde el principio
    usando <strong>Named Aggregation</strong>.
  </div>

  <p style="font-size:12px;font-weight:600;letter-spacing:.8px;text-transform:uppercase;color:#475569;margin-bottom:.6rem;">
    Opción A — Aplanar con list comprehension (clásica)
  </p>

  <div class="code-block">
    <div class="code-header">
      <div class="dots"><div class="dot dot-r"></div><div class="dot dot-y"></div><div class="dot dot-g"></div></div>
      <span class="code-lang">Python</span>
    </div>
    <pre class="cc"><span class="al">df5</span> <span class="op">=</span> df.groupby(<span class="st">"Continente"</span>).agg({
    <span class="num">2010</span>: [<span class="st">"min"</span>, <span class="st">"max"</span>, <span class="st">"sum"</span>],
    <span class="num">2011</span>: [<span class="st">"min"</span>, <span class="st">"max"</span>],
    <span class="num">2012</span>: [<span class="st">"sum"</span>]
})

<span class="cm"># El resultado tiene multi-índice: (2010, "min"), (2010, "max")…</span>
<span class="fn">print</span>(<span class="al">df5</span>.columns)

<span class="cm"># Aplanar con list comprehension</span>
<span class="al">df5</span>.columns <span class="op">=</span> [<span class="fn">f</span><span class="st">"{x[0]}_{x[1]}"</span> <span class="kw">for</span> x <span class="kw">in</span> <span class="al">df5</span>.columns.values]
<span class="al">df5</span>.reset_index()</pre>
  </div>

  <hr class="nb-divider-sm">

  <p style="font-size:12px;font-weight:600;letter-spacing:.8px;text-transform:uppercase;color:#475569;margin-bottom:.6rem;">
    Opción B — Named Aggregation (moderna, sin multi-índice)
  </p>

  <div class="code-block">
    <div class="code-header">
      <div class="dots"><div class="dot dot-r"></div><div class="dot dot-y"></div><div class="dot dot-g"></div></div>
      <span class="code-lang">Python</span>
    </div>
    <pre class="cc"><span class="cm"># Cada argumento define directamente el nombre de la columna resultante</span>
df.groupby(<span class="st">"Continente"</span>).agg(
    min_2010 <span class="op">=</span> (<span class="num">2010</span>, <span class="st">"min"</span>),
    max_2010 <span class="op">=</span> (<span class="num">2010</span>, <span class="st">"max"</span>),
    sum_2010 <span class="op">=</span> (<span class="num">2010</span>, <span class="st">"sum"</span>),
    min_2011 <span class="op">=</span> (<span class="num">2011</span>, <span class="st">"min"</span>),
    max_2011 <span class="op">=</span> (<span class="num">2011</span>, <span class="st">"max"</span>),
    sum_2012 <span class="op">=</span> (<span class="num">2012</span>, <span class="st">"sum"</span>)
)</pre>
  </div>

  <div class="nb-note">
    <i class="ti ti-info-circle" style="color:#0F766E;font-size:15px;flex-shrink:0;margin-top:1px;"></i>
    <span class="nb-note-text">
      La <strong>Opción B</strong> es la forma recomendada en pandas moderno — evita el multi-índice
      desde el principio y produce columnas con nombres legibles directamente, sin necesitar
      list comprehension posterior. La sintaxis es
      <code class="ic">nombre_columna = (columna_original, "función")</code>.
    </span>
  </div>

  <div class="warn-box">
    <i class="ti ti-alert-triangle" style="color:#D97706;font-size:15px;flex-shrink:0;margin-top:1px;"></i>
    <span class="warn-text">
      Usa siempre las funciones como <strong>strings</strong>:
      <code class="ic">"min"</code>, <code class="ic">"max"</code>, <code class="ic">"sum"</code>…
      Pasar las funciones de Python directamente (<code class="ic">min</code>, <code class="ic">max</code>)
      genera un <strong>FutureWarning</strong> en pandas 3.x.
    </span>
  </div>

</div>
```
