```python
# ═══════════════════════════════════════════════════════════════════════════
# EJERCICIO 1 — plot() + scatter() en la misma figura + personalización
# Objetivo: ver cómo coexisten línea y puntos, y practicar parámetros de estilo
# ═══════════════════════════════════════════════════════════════════════════

x = np.linspace(0, 2 * np.pi, 100)   # 100 puntos entre 0 y 2π

# ── Tamaño del lienzo ────────────────────────────────────────────────────────
# PRUEBA: cambia (10, 4) a (6, 6) o (14, 3) y observa cómo cambia la proporción
plt.figure(figsize=(10, 4))

# ── plt.plot() — línea ───────────────────────────────────────────────────────
# PRUEBA: cambia ls='--' a '-', '-.', ':' para ver distintos estilos de línea
# PRUEBA: cambia lw=2 a 0.5 o 4 para ver el grosor
plt.plot(x, np.sin(x),
         color='royalblue', ls='--', lw=2,
         marker='o', ms=4,             # marcador circular de tamaño 4
         label='sen(x)')

# ── segunda curva en la misma figura ─────────────────────────────────────────
# PRUEBA: añade alpha=0.4 y observa la transparencia
plt.plot(x, np.cos(x),
         color='tomato', ls='-', lw=1.5,
         label='cos(x)')

# ── plt.scatter() — nube de puntos ───────────────────────────────────────────
# PRUEBA: cambia s=20 a 100 para ver puntos más grandes
# PRUEBA: cambia alpha=0.5 a 1.0 o 0.1
np.random.seed(42)
xs = np.random.uniform(0, 2*np.pi, 30)
plt.scatter(xs, np.sin(xs) + np.random.normal(0, 0.15, 30),
            s=20, color='navy', alpha=0.5, zorder=5,  # zorder pone los puntos encima
            label='datos con ruido')

# ── Personalización ──────────────────────────────────────────────────────────
plt.title('plot() y scatter() combinados', fontsize=14, fontweight='bold')
plt.xlabel('x (radianes)')
plt.ylabel('Amplitud')
plt.legend()    # muestra los labels definidos arriba
plt.grid(True, linestyle=':', alpha=0.4)   # PRUEBA: alpha=1.0 o grid(False)

plt.tight_layout()   # evita que los elementos se corten
plt.show()
```

```python
# ═══════════════════════════════════════════════════════════════════════════
# EJERCICIO 2 — subplots 2×2 con bar, hist, pie y scatter3D (dataset iris)
# Muestra en un único bloque: barras, histograma, sectores y gráfico 3D
# ═══════════════════════════════════════════════════════════════════════════

# ── Preparar datos ───────────────────────────────────────────────────────────
medias = iris.groupby('species')['petal_length'].mean()  # media del pétalo por especie
conteos = iris['species'].value_counts()                 # 50 flores por especie

# PRUEBA: cambia figsize=(14, 10) a (10, 7) y observa cómo se compacta la cuadrícula
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Matplotlib — cuatro tipos de gráfico con iris', fontsize=15, fontweight='bold')

# ── Subplot [0,0]: Barras verticales ─────────────────────────────────────────
# PRUEBA: cambia colors a ['#1E3A8A','#0F766E','#DC2626']
colors_esp = ['steelblue', 'salmon', 'mediumseagreen']
axes[0, 0].bar(medias.index, medias.values,
               color=colors_esp, edgecolor='white', width=0.5)
axes[0, 0].set_title('Longitud media del pétalo por especie')
axes[0, 0].set_ylabel('cm')

# ── Subplot [0,1]: Histograma superpuesto por especie ────────────────────────
# PRUEBA: cambia bins=15 a 5 o 30 para ver más o menos detalle
for especie, color in zip(iris['species'].unique(), colors_esp):
    datos = iris[iris['species'] == especie]['sepal_length']
    axes[0, 1].hist(datos, bins=15, alpha=0.5, color=color, label=especie)
axes[0, 1].set_title('Distribución de sepal_length')
axes[0, 1].set_xlabel('cm')  ;  axes[0, 1].legend()

# ── Subplot [1,0]: Pie chart (sectores) ──────────────────────────────────────
# PRUEBA: cambia explode a (0.05, 0.05, 0.05) para separar un poco los sectores
axes[1, 0].pie(conteos.values,
               labels=conteos.index,
               colors=colors_esp,
               autopct='%1.1f%%',   # muestra el porcentaje con 1 decimal
               startangle=90,       # PRUEBA: cambia a 0 o 180
               shadow=True)
axes[1, 0].set_title('Proporción de especies (pie chart)')

# ── Subplot [1,1]: Scatter 3D ─────────────────────────────────────────────────
# El subplot 3D hay que crearlo explícitamente después de definir la figura
axes[1, 1].remove()   # eliminamos el eje 2D que se creó automáticamente
ax3d = fig.add_subplot(2, 2, 4, projection='3d')  # lo reemplazamos con uno 3D

for especie, color in zip(iris['species'].unique(), colors_esp):
    sub = iris[iris['species'] == especie]
    ax3d.scatter(sub['sepal_length'], sub['sepal_width'], sub['petal_length'],
                 c=color, label=especie, s=30, alpha=0.7)
ax3d.set_xlabel('sepal_length') ; ax3d.set_ylabel('sepal_width') ; ax3d.set_zlabel('petal_length')
ax3d.set_title('Scatter 3D — Iris')
ax3d.legend(fontsize=8)
# PRUEBA: descomenta la siguiente línea para rotar la vista del 3D
# ax3d.view_init(elev=20, azim=60)

plt.tight_layout()
# Si quisieras guardar: plt.savefig('figura.png', dpi=300, bbox_inches='tight') — ANTES de show()
plt.show()
```

```python
# ═══════════════════════════════════════════════════════════════════════════
# EJERCICIO 4 — kdeplot, boxplot, violinplot, heatmap, radar chart
# Todos en un único bloque para maximizar el contenido cubierto
# ═══════════════════════════════════════════════════════════════════════════

# ── 4a: kdeplot + boxplot (comparación Matplotlib vs Seaborn) ────────────────
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

# Boxplot matplotlib puro
datos_bp = [iris[iris['species'] == sp]['sepal_length'].values
            for sp in iris['species'].unique()]
axes[0].boxplot(datos_bp, labels=iris['species'].unique(), patch_artist=True,
                boxprops=dict(facecolor='steelblue', alpha=0.6))
axes[0].set_title('Matplotlib — boxplot()')  ;  axes[0].set_ylabel('sepal_length')

# Boxplot Seaborn (una línea, colores por hue automáticos)
# PRUEBA: cambia palette='Set2' a 'pastel', 'deep', 'muted'
sns.boxplot(data=iris, x='species', y='sepal_length',
            hue='species', palette='Set2', ax=axes[1], legend=False)
axes[1].set_title('Seaborn — boxplot()')

# Violinplot Seaborn — muestra la forma de la distribución
# PRUEBA: cambia inner='box' a 'quartile', 'point', None
sns.violinplot(data=iris, x='species', y='sepal_length',
               hue='species', palette='Set1', inner='box', ax=axes[2], legend=False)
axes[2].set_title('Seaborn — violinplot()')

plt.suptitle('Boxplot (Matplotlib vs Seaborn) + Violinplot', fontsize=13, fontweight='bold')
plt.tight_layout()  ;  plt.show()

# ── 4b: KDE plot + heatmap de correlación ─────────────────────────────────────
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# KDE por especie — PRUEBA: cambia fill=True a False para ver solo la curva
# PRUEBA: cambia bw_adjust=1.0 a 0.4 (más suave) o 2.0 (más suave aún)
for especie, color in zip(iris['species'].unique(), ['steelblue','salmon','seagreen']):
    sns.kdeplot(data=iris[iris['species'] == especie], x='petal_length',
                fill=True, alpha=0.4, color=color, label=especie, bw_adjust=1.0, ax=ax1)
ax1.set_title('kdeplot — distribución de petal_length')  ;  ax1.legend()

# Heatmap de correlación
corr = iris.select_dtypes('number').corr()   # solo columnas numéricas
# PRUEBA: cambia cmap='coolwarm' a 'viridis', 'RdBu', 'Blues'
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm',
            linewidths=0.5, ax=ax2,
            vmin=-1, vmax=1)   # PRUEBA: elimina vmin/vmax y observa el cambio de escala
ax2.set_title('Heatmap — correlación Iris')

plt.tight_layout()  ;  plt.show()

# ── 4c: Radar chart (solo con Matplotlib — proyección polar) ──────────────────
# Seaborn no tiene radar nativo; se usa plt con projection='polar'
labels  = np.array(iris.columns[:-1])   # 4 variables numéricas
angles  = np.linspace(0, 2*np.pi, len(labels), endpoint=False)

fig = plt.figure(figsize=(6, 6))
ax  = fig.add_subplot(polar=True)

# Dibuja 3 flores de distintas especies para comparar
# PRUEBA: cambia los índices 0, 50, 100 a otras filas del DataFrame
colores_radar = ['royalblue', 'tomato', 'mediumseagreen']
for idx, color in zip([0, 50, 100], colores_radar):
    pattern = iris.iloc[idx, :-1].values
    especie = iris.iloc[idx, -1]
    pattern = np.concatenate((pattern, [pattern[0]]))   # cerrar el polígono
    ang = np.concatenate((angles, [angles[0]]))
    ax.plot(ang, pattern, 'o-', lw=2, color=color, label=especie)
    ax.fill(ang, pattern, alpha=0.15, color=color)

ax.set_thetagrids(angles * 180/np.pi, labels)  # coloca etiquetas en los ángulos
ax.set_title('Radar chart — comparación de 3 flores', pad=15)
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
plt.show()
```
